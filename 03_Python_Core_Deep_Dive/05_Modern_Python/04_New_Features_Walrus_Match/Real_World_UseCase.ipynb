# Real World Use Case: Parsers (CLI Tools)

**Scenario**: Building a Command Line Interface (CLI).
**Pattern Matching**: Perfect for parsing user commands.
```python
command = input("> ").split()
match command:
    case ["quit"]:
        break
    case ["load", filename]:
        load(filename)
    case ["save", filename]:
        save(filename)
    case _:
        print("Invalid command")
```
**Walrus**: Useful when reading files in chunks.
```python
while (chunk := file.read(1024)):
    process(chunk)
```
