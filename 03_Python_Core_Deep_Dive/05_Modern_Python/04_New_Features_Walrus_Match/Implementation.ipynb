# Implementation: Walrus and Match

**Goal**: Conciseness.

```python
# 1. Walrus Operator
data = [1, 2, 3, 4, 15]

# Old way:
# n = len(data)
# if n > 10: ...

# New way:
if (n := len(data)) > 10:
    print(f"List is too long ({n} elements)")
else:
    print(f"List is okay ({n} elements)")

# 2. Pattern Matching
def http_status(status):
    match status:
        case 200:
            return "OK"
        case 400 | 404:
            return "Not Found"
        case 500:
            return "Server Error"
        case _:
            return "Unknown"

print(http_status(404))
```
