# Implementation: Strong Typing

**Goal**: Clarity.

```python
from typing import List, Optional

def process_scores(scores: List[float], bonus: bool = False) -> Optional[float]:
    if not scores:
        return None
    
    avg = sum(scores) / len(scores)
    if bonus:
        avg += 10
    return avg

result = process_scores([80.5, 90.0], bonus=True)
print(f"Average: {result}")

# result = process_scores(["A", "B"])  # This runs but would fail mypy check
print("Type hints are documentation that can be tested.")
```
