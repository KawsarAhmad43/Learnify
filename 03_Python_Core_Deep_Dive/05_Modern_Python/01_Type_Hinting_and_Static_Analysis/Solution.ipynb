# Solution: Overloads

**Objective**: Multiple signatures.
1. **Problem**: A function behaves differently based on input types.
   *   `get(x: int) -> str`
   *   `get(x: str) -> int`
2. **Solution**: Use `@overload` decorator multiple times to define signatures, then one final implementation without the decorator.
   ```python
   @overload
   def get(x: int) -> str: ...
   @overload
   def get(x: str) -> int: ...
   def get(x):
       # Implementation
   ```
