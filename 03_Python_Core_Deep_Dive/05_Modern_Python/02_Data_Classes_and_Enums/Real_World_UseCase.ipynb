# Real World Use Case: Configuration Management

**Scenario**: Machine Learning Hyperparameters.
**Old Way**: A massive dictionary `config = {'lr': 0.01, 'batch': 32}`. Prone to typos (`config['lR']`).
**New Way**: 
```python
@dataclass
class Config:
    learning_rate: float = 0.01
    batch_size: int = 32
```
**Library**: `Hydra` uses dataclasses to manage complex hierarchical configurations for PyTorch experiments.
