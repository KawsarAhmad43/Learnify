# Implementation: Inventory System

**Goal**: Clean Code.

```python
from dataclasses import dataclass, field
from enum import Enum, auto
from typing import List

class ItemType(Enum):
    WEAPON = auto()
    ARMOR = auto()
    POTION = auto()

@dataclass
class Item:
    name: str
    price: float
    type: ItemType
    # Default factory for mutable defaults (List/Dict)
    tags: List[str] = field(default_factory=list)

sword = Item("Excalibur", 1000.0, ItemType.WEAPON, ["Legendary"])
shield = Item("Wooden Shield", 50.0, ItemType.ARMOR)

print(sword)
# Output: Item(name='Excalibur', price=1000.0, type=<ItemType.WEAPON: 1>, tags=['Legendary'])
# Note: __repr__ was auto-generated!
```
