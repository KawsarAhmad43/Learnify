# Solution: Multiprocessing

**Objective**: CPU Heavy Tasks.
1. **GIL**: Global Interpreter Lock prevents threads from using multiple CPU cores.
2. **AsyncIO**: Also limited to 1 core (Single Threaded).
3. **Multiprocessing**: Spawns separate Python processes. Each has its own GIL. True Parallelism.
   `from multiprocessing import Pool`
