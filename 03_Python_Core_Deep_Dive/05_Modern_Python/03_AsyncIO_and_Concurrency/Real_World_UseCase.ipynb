# Real World Use Case: Web Servers (Node.js style)

**Scenario**: Handling 10,000 requests/sec.
**Thread-based (Flask)**: Each request spawns a thread. Threads have overhead (memory stack). The OS struggles with 10k threads.
**Async-based (FastAPI/Uvicorn)**: Use 1 thread. When a request waits for DB, Python switches to handle the next request.
**Result**: Massive throughput with minimal resources.
