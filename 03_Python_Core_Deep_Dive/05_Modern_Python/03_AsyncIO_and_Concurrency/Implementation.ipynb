# Implementation: Async Download

**Goal**: Speed.

```python
import asyncio
import time

async def mock_download(id, duration):
    print(f"Start download {id}...")
    await asyncio.sleep(duration)
    print(f"Finished download {id}.")
    return duration

async def main():
    start = time.time()
    
    # Run 3 downloads concurrently
    # If sequential: 1 + 2 + 1 = 4 seconds
    # If concurrent: Max(1, 2, 1) = 2 seconds
    results = await asyncio.gather(
        mock_download("A", 1),
        mock_download("B", 2),
        mock_download("C", 1)
    )
    
    end = time.time()
    print(f"Total time: {end - start:.2f} seconds")

# Note: In Jupyter, asyncio is already running. You might need just 'await main()' instead of 'asyncio.run(main())'
# For script usage:
# asyncio.run(main()) 
print("AsyncIO is crucial for web scraping and API servers.")
```
