# Solution: Context Managers

**Objective**: `with` statement.
1. **Magic**: `__enter__` and `__exit__`.
2. **Use Case**: Determining duration of execution.
   ```python
   with Timer():
       heavy_process()
   ```
3. **Implementation**:
   *   `__enter__`: Start clock. Return self.
   *   `__exit__`: Stop clock. Print time difference.
