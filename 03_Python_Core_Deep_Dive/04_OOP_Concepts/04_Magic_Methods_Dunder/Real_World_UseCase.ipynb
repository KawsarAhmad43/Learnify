# Real World Use Case: PyTorch Tensors

**Scenario**: Deep Learning.
**Magic**: When you write `loss = (y_pred - y_true) ** 2`, you are NOT subtracting numbers. You are calling `y_pred.__sub__(y_true)` and then `__pow__(2)`.
**Under the hood**: These magic methods trigger complex C++/CUDA kernels (CuDNN) to run the math on the GPU.
**Result**: Python syntax, C speed.
