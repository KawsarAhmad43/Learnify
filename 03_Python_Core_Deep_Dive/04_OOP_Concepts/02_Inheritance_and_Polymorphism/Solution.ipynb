# Solution: Composition over Inheritance

**Objective**: Design Principle.
1. **Inheritance**: 'Is-A'. (Tiger is a Cat).
2. **Composition**: 'Has-A'. (Car has an Engine).
3. **Advice**: Prefer Composition. Inheritance creates tight coupling. If you change the Parent, you break all Children. Composition is more flexible.
