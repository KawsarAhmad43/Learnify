# Implementation: Animal Kingdom

**Goal**: Basic OOP Structure.

```python
class Animal:
    kingdom = "Animalia" # Class Attribute

    def __init__(self, name, age):
        self.name = name # Instance Attribute
        self.age = age

    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

d = Dog("Buddy", 5)
c = Cat("Whiskers", 3)

print(f"{d.name} says {d.speak()}")
print(f"{c.name} says {c.speak()}")
print(f"They both belong to {d.kingdom}")
```
