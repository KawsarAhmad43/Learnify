# Solution: Object Lifecycle

**Objective**: Memory.
1. **Creation**: `__new__` (Allocates memory) -> `__init__` (Initializes attributes).
2. **Usage**: Methods are called.
3. **Destruction**: `__del__` (Called when ref count hits 0). Rarely used in Python because Garbage Collection handles it, but good to know for managing external resources (files/sockets).
