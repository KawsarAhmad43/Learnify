# Implementation: Measuring Disparate Impact

**Goal**: Parity Check.

```python
import pandas as pd

# 1. Mock Data (Loan Approvals)
data = pd.DataFrame({
    "Group": ["A", "A", "A", "A", "B", "B", "B", "B"],
    "Outcome": [1, 1, 1, 0, 0, 1, 0, 0] # 1=Approved, 0=Denied
})

# 2. Calculate Approval Rates
rate_a = data[data["Group"] == "A"]["Outcome"].mean()
rate_b = data[data["Group"] == "B"]["Outcome"].mean()

print(f"Approval Rate A: {rate_a * 100}%")
print(f"Approval Rate B: {rate_b * 100}%")

# 3. Disparate Impact Ratio (Standard is 80% Rule)
di_ratio = rate_b / rate_a
print(f"DI Ratio: {di_ratio:.2f}")

if di_ratio < 0.8:
    print("FAIL: Significant adverse impact detected.")
else:
    print("PASS: Within acceptable limits.")
```

## Conclusion
Simple math reveals systemic issues.
