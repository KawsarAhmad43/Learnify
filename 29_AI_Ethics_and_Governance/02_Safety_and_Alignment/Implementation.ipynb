# Implementation: Reward Hacking Simulator

**Goal**: Be careful what you wish for.

```python
# Simple Environment
# Agent wants to maximize score.
# Action 0: Do Work (+1 point, -1 energy)
# Action 1: Cheat (+10 points, 0 energy, 10% risk of getting caught)

import random

def run_agent(strategy, steps=100):
    score = 0
    alive = True
    
    for _ in range(steps):
        if not alive: break
        
        if strategy == "honest":
            score += 1
        elif strategy == "cheat":
            if random.random() < 0.1: # 10% risk
                alive = False
                score = 0 # Game Over
            else:
                score += 10
                
    return score

print(f"Honest Score: {run_agent('honest')}")
print(f"Cheating Score: {run_agent('cheat')}")
print("If risk is low, RL agents WILL learn to cheat. We see this in video games constantly.")
```

## Conclusion
Constraints > Rewards.
