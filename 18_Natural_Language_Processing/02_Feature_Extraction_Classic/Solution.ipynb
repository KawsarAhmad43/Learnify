# Solution: TF-IDF Math

**Objective**: Verify the "Zero Score" for common words.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    "cat dog",
    "cat cat"
]

# 1. Standard Sklearn (Smooth IDF is on by default)
vectorizer = TfidfVectorizer(norm=None) # Turn off normalization to see raw scores
X = vectorizer.fit_transform(corpus)

feature_names = vectorizer.get_feature_names_out()
print("Features:", feature_names)
print("Vectors:\n", X.todense())

# Notice 'cat' is not 0 because Sklearn uses log(N+1 / df+1) + 1 to avoid division by zero errors.
# But 'dog' should be higher than 'cat' in document 1.
```
