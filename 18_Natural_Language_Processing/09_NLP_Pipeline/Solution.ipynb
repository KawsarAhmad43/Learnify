# Solution: Truncation

**Objective**: Implement truncation logic.

```python
def truncate(seq, strategy, max_len=5):
    if len(seq) <= max_len:
        return seq
    
    if strategy == "head":
        return seq[:max_len]
    elif strategy == "tail":
        return seq[-max_len:]
    elif strategy == "head_tail":
        head_len = max_len // 2
        tail_len = max_len - head_len
        return seq[:head_len] + seq[-tail_len:]

seq = [1, 2, 3, 4, 5, 6, 7]

print(f"Original: {seq}")
print(f"Head: {truncate(seq, 'head')}")
print(f"Tail: {truncate(seq, 'tail')}")
print(f"Mix: {truncate(seq, 'head_tail')}")
```
