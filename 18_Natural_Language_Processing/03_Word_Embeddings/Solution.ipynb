# Solution: Odd One Out

**Objective**: Detect outliers.

```python
import numpy as np

data = {
    "apple": np.array([1.0, 0.0]),
    "banana": np.array([0.9, 0.1]),
    "orange": np.array([0.95, 0.05]),
    "car": np.array([0.0, 1.0])
}

# 1. Centroid
vectors = np.array(list(data.values()))
centroid = np.mean(vectors, axis=0)
print(f"Centroid: {centroid}")

# 2. Distances
print("Distances from Centroid:")
max_dist = -1
outlier = None

for word, vec in data.items():
    dist = np.linalg.norm(vec - centroid)
    print(f"{word}: {dist:.4f}")
    if dist > max_dist:
        max_dist = dist
        outlier = word

print(f"\nOdd one out: {outlier.upper()}")
```
