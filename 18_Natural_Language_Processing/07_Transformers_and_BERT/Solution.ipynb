# Solution: Softmax

**Objective**: Pick the winner.

```python
import numpy as np

logits = {
    "cat": 2.5,
    "dog": 2.1,
    "fox": 5.8,
    "table": -1.0
}

def softmax(logits_dict):
    values = np.array(list(logits_dict.values()))
    keys = list(logits_dict.keys())
    
    # Exp
    exps = np.exp(values)
    # Normalize
    probs = exps / np.sum(exps)
    
    return dict(zip(keys, probs))

probabilities = softmax(logits)
print("Probabilities:")
for k, v in probabilities.items():
    print(f"{k}: {v:.4f}")
    
print(f"\nWinner: {max(probabilities, key=probabilities.get).upper()}")
```
