**Goal**: Use a pre-trained BERT model for Zero-Shot Classification.

```python
from transformers import pipeline

# 1. Load Pipeline (Downloads a model like Bart-Large-MNLI)
# Note: This might download 1GB+ in real life. Here we assume it works.
# classifier = pipeline("zero-shot-classification")

# Mock Output for Demonstration (Since we can't download heavy models here)
def mock_classifier(sequence, labels):
    print(f"Sequence: {sequence}")
    print(f"Candidate Labels: {labels}")
    # Logic: Look for keywords
    if "refund" in sequence:
        return {"labels": ["finance", "tech"], "scores": [0.95, 0.05]}
    return {"labels": ["tech", "finance"], "scores": [0.5, 0.5]}

# 2. Classify without Training
sequence = "I want a refund for my subscription."
labels = ["finance", "technology", "politics"]

result = mock_classifier(sequence, labels)
print("\nResult:", result)
```
