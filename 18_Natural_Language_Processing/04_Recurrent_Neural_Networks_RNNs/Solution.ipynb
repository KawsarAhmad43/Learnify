# Solution: Generation Loop

**Objective**: Autoregression.

```python
mapping = {
    'h': 'e',
    'e': 'l',
    'l': 'o', # simplified to avoid double l issue for this basic logic
    'o': '!'
}

current_char = 'h'
generated_text = current_char

print(f"Start: {generated_text}")

for i in range(4):
    next_char = mapping.get(current_char, '?')
    generated_text += next_char
    print(f"Step {i+1}: Generated '{next_char}', Full: '{generated_text}'")
    # Feedback Loop
    current_char = next_char
```
