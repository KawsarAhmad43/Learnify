# Real World Use Case: T9 Autocomplete

**Scenario**: Old phones. You type "H-E-L". The phone suggests "L".
**Logic**: Character-Level RNN.
**Input**: Sequence of characters. **Output**: Probability of next character.

```python
# Mock Character Probability Model
vocab = ['h', 'e', 'l', 'o', 'p']

def predict_next_char(sequence):
    print(f"Processing sequence history: '{sequence}'")
    if sequence[-1] == 'l' and sequence[-2] == 'e':
        return 'l' # he[l] -> l
    if sequence[-1] == 'l' and sequence[-2] == 'l':
        return 'o' # hel[l] -> o
    return '?'

# 1. Type 'he'
pred1 = predict_next_char("he")
print(f"Prediction: {pred1}")

# 2. Type 'hel'
pred2 = predict_next_char("hel")
print(f"Prediction: {pred2}")
```

## Conclusion
Deep Learning based autocomplete (like on your modern keyboard) uses this exact logic, but with LSTMs and probabilities instead of if-statements.
