# Solution: Stems vs Lemmas

**Objective**: Compare NLTK tools.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

words = ["running", "flies", "better", "meeting"]

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print(f"{'Word':<10} | {'Stem':<10} | {'Lemma (Noun)':<15} | {'Lemma (Verb)':<15}")
print("-"*60)

for w in words:
    stem = stemmer.stem(w)
    lemma_n = lemmatizer.lemmatize(w, pos='n') # Default
    lemma_v = lemmatizer.lemmatize(w, pos='v') # Explicit verb
    print(f"{w:<10} | {stem:<10} | {lemma_n:<15} | {lemma_v:<15}")
```
