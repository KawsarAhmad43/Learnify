# Implementation: Text Generation (GPT-2)

**Goal**: Use HuggingFace to generate text.

```python
from transformers import pipeline, set_seed

# 1. Load Generator (Downloads small GPT-2)
# generator = pipeline('text-generation', model='gpt2')
set_seed(42)

# Mock Output for Demonstration
class MockGenerator:
    def __call__(self, prompt, max_length, num_return_sequences):
        # Simple rule-based completion to simulate AI
        continuations = [
            " is the most important technology of our time. It will change everything.",
            " is dangerous if not controlled. We need safety guidelines."
        ]
        return [{'generated_text': prompt + c} for c in continuations]

generator = MockGenerator()

# 2. Generate
prompt = "Artificial Intelligence"
results = generator(prompt, max_length=30, num_return_sequences=2)

for i, res in enumerate(results):
    print(f"--- Choice {i+1} ---")
    print(res['generated_text'])
```
