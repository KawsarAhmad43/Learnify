# Solution: Database Query

**Objective**: Attention as Retrieval.

```python
import numpy as np

# 1. Embeddings (Orthogonal)
keys = {
    "color": np.array([1, 0, 0]),
    "animal": np.array([0, 1, 0]),
    "fruit": np.array([0, 0, 1])
}

values = {
    "color": "red",
    "animal": "cat",
    "fruit": "apple"
}

query = np.array([0, 0, 1]) # Looking for "fruit"

# 2. Attention Scores
scores = {}
for k_str, k_vec in keys.items():
    score = np.dot(query, k_vec)
    scores[k_str] = score
    print(f"Score for {k_str}: {score}")

# 3. Softmax (Hard Logic here for simplicity)
best_key = max(scores, key=scores.get)
print(f"\nBest Match: {best_key}")
print(f"Retrieved Value: {values[best_key]}")
```
