# Solution: Memory Decay

**Objective**: Visualize Forget Gate dynamics.

```python
memory = 100.0
forget_rate = 0.5

print(f"Start Memory: {memory}")

for t in range(1, 6):
    memory = memory * forget_rate
    print(f"Step {t}: Memory {memory}")

# It decays exponentially: 100 -> 50 -> 25 -> 12.5 -> 6.25
```
