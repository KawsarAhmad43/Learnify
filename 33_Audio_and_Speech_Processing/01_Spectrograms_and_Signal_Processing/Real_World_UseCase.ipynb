# Real World Use Case: Shazam

**Scenario**: Identifying a song in a noisy bar.
**Method**: Audio Fingerprinting.
**Logic**:
1.  Convert Song to Spectrogram.
2.  Find "Peaks" (High intensity points).
3.  Create a Hash of local constellations of peaks (e.g., "Peak A at 100Hz, Peak B at 300Hz, 50ms apart").
4.  Match hashes against a massive database.
**Result**: Identifies the song in 3 seconds, even with background noise, because the *relative* position of peaks remains constant.
