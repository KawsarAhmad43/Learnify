# Implementation: pyttsx3 (Offline)

**Goal**: Synthesis.

```python
import pyttsx3

# 1. Initialize Engine
engine = pyttsx3.init()

# 2. Properties
engine.setProperty('rate', 150)    # Speed
engine.setProperty('volume', 0.9)  # Volume

# 3. Speak
engine.say("I am a robot, but I am learning to sing.")
engine.runAndWait()
print("For Neural TTS, use 'Bark' or 'Tortoise-TTS' (Requires GPU).")
```

## Conclusion
Offline TTS is fast but low quality.
