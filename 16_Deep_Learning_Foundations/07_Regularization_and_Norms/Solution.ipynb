# Solution: L2 Regularization Calculation

**Objective**: Calculate Total Loss.

```python
import numpy as np

mse = 0.5
weights = np.array([0.1, 4.0, -0.5, 0.0])
lmbda = 0.1

# 1. Sum of Squares
sum_sq = np.sum(weights ** 2)
print(f"Sum of Squares: {sum_sq}")
# 0.01 + 16 + 0.25 + 0 = 16.26

# 2. Penalty
penalty = lmbda * sum_sq
print(f"Penalty term: {penalty}")
# 1.626

# 3. Total Loss
total_loss = mse + penalty
print(f"Total Loss: {total_loss}")
```

## Conclusion
The loss jumped from 0.5 to 2.126 because of the large weight (4.0). 
The model is now "unhappy" about that large weight and will reduce it.
