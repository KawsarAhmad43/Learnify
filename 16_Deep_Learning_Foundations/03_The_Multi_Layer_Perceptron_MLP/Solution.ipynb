# Solution: Manual Forward Pass

**Objective**: Verify manual calculations with NumPy.

```python
import numpy as np

# 1. Setup
x = np.array([1, 2])

W1 = np.array([
    [1, 1],   # Input 1 weights to Node 1, Node 2
    [-2, 1]   # Input 2 weights to Node 1, Node 2
]) # Shape (2, 2)

b1 = np.zeros(2)

W2 = np.array([
    [1],
    [-1]
]) # Shape (2, 1)

b2 = np.zeros(1)

# 2. Forward Pass

# Layer 1
z1 = np.dot(x, W1) + b1 
# Calculation:
# Node 1: 1*1 + 2*(-2) = 1 - 4 = -3
# Node 2: 1*1 + 2*1 = 1 + 2 = 3
# z1 = [-3, 3]
print(f"z1: {z1}")

a1 = np.maximum(0, z1)
# ReLU([-3, 3]) -> [0, 3]
print(f"a1: {a1}")

# Layer 2
output = np.dot(a1, W2) + b2
# Calculation:
# 0*1 + 3*(-1) = -3
print(f"Final Output: {output}")
```
