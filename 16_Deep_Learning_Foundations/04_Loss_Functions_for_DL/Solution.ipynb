# Solution: Manual Loss Calculation

**Objective**: Verify manual calculations.

```python
import numpy as np

# Case 1: Bad Prediction
# y=1, y_hat=0.2
loss_bad = -1 * np.log(0.2)
print(f"Loss for Bad Prediction (0.2): {loss_bad:.4f}")

# Case 2: Good Prediction
# y=1, y_hat=0.9
loss_good = -1 * np.log(0.9)
print(f"Loss for Good Prediction (0.9): {loss_good:.4f}")
```

## Conclusion
The bad prediction has a loss of **1.61**. The good prediction has a loss of **0.10**.
The optimizer will try to move weights to get from 1.61 down to 0.10.
