# Solution: Gradient Calculation

 Function: $f = xy + ReLU(x)$

```python
def calculate_gradients(x, y):
    # Forward
    relu_x = max(0, x)
    f = (x * y) + relu_x
    print(f"--- Input x={x}, y={y} ---")
    print(f"Forward f: {f}")
    
    # Backward wrt x
    # deriv(x*y) = y
    # deriv(ReLU(x)) = 1 if x>0 else 0
    grad_relu = 1 if x > 0 else 0
    df_dx = y + grad_relu
    
    print(f"df/dx: {df_dx}")
    return df_dx

# Case 1: x=3
calculate_gradients(3, -2)
# Check: -2 + 1 = -1

# Case 2: x=-3
calculate_gradients(-3, -2)
# Check: -2 + 0 = -2
```
