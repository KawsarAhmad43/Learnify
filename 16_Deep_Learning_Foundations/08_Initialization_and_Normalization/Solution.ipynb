# Solution: Standardization

**Objective**: Z-Score Normalization.

```python
import numpy as np

data = np.array([100, 200, 300])

# 1. Calculate Statistics
mean = np.mean(data)
std = np.std(data)

print(f"Mean: {mean}")
print(f"Std: {std}")

# 2. Normalize
data_norm = (data - mean) / std

print(f"Normalized Data: {data_norm}")

# 3. Verify
print(f"New Mean: {np.mean(data_norm):.2f}")
print(f"New Std: {np.std(data_norm):.2f}")
```
