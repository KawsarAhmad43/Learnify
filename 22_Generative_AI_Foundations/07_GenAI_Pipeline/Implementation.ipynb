# Implementation: Rejection Sampling

**Goal**: Filter bad outputs.

```python
import numpy as np

# 1. Mock Generator (Produces values 0-100)
def generate_batch(n):
    return np.random.randint(0, 100, n)

# 2. Mock Discriminator (Reward Model)
# Prefers high numbers
def score(samples):
    return samples # The value itself is the score

# 3. Pipeline
N = 10
candidates = generate_batch(N)
scores = score(candidates)

best_idx = np.argmax(scores)
best_sample = candidates[best_idx]

print(f"Candidates: {candidates}")
print(f"Chosen: {best_sample}")
print("We traded Compute (10x cost) for Quality (Max of 10).")
```

This is how 'DeepSeek R1' or 'O1' works (Thinking time = Generating many thoughts and filtering).
