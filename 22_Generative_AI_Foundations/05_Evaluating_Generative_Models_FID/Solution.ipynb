# Solution: Statistical Distances

**Objective**: Intuition check.

```python
import numpy as np

real_colors = np.array([0, 100, 200]) # Diverse
fake_colors_collapsed = np.array([100, 100, 100]) # Mean is 100, but Variance is 0

real_std = np.std(real_colors)
fake_std = np.std(fake_colors_collapsed)

print(f"Real Std: {real_std:.2f}")
print(f"Fake Std: {fake_std:.2f}")
print("Difference in Std leads to high FID.")
```
