# Real World Use Case: Weather Forecasting (GraphCast)

**Problem**: Predicting rain.
**Why Flows?**: We need to know the *exact probability* of extreme events (e.g., Cat 5 Hurricane).
*   GANs might just output "Cloudy" (Mode Average).
*   Flows give us a precise likelihood map: P(Rain=100mm) = 0.0001.
**Method**: Train a Flow model on global atmospheric states.
