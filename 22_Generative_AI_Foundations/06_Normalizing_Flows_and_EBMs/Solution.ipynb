# Solution: Energy

**Objective**: Boltzmann calculation.

```python
import math

energy_valid = 0
energy_invalid = 10

prob_valid = math.exp(-energy_valid)
prob_invalid = math.exp(-energy_invalid)

print(f"Prob Valid: {prob_valid}")
print(f"Prob Invalid: {prob_invalid:.6f}")
print("The EBM pushes inputs towards valid configurations.")
```
