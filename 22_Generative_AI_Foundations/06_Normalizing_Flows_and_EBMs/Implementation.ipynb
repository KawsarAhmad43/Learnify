# Implementation: Jacobian Volume Change

**Goal**: Understand density transformation.

```python
import numpy as np

# 1. Original Distribution (Uniform 0-1)
# Probability of picking 0.5 is 1.0

# 2. Function f(x) = 2x
# This stretches the domain from [0, 1] to [0, 2].
# The total probability must sum to 1.
# So the new density must be 0.5 everywhere.

# 3. Jacobian Calculation
# f'(x) = 2.
# determinant = 2.
# P(y) = P(x) / det = 1.0 / 2 = 0.5.

def flow_metrics(scale_factor):
    det = scale_factor
    density_change = 1 / det
    return density_change

print(f"Scale 2x -> New Density: {flow_metrics(2)}")
print(f"Scale 0.5x -> New Density: {flow_metrics(0.5)} (Concentrated)")
```

## Conclusion
Flows are just chains of these transformations.
