# Solution: Bayes Rule

**Objective**: Deriving classification from generation.

```python
# Mock Probabilities
p_cat = 0.5
p_image_given_cat = 0.8
p_image = 0.5 # Normalizing constant

p_cat_given_image = (p_image_given_cat * p_cat) / p_image
print(f"P(Cat|Image): {p_cat_given_image:.2f}")

# This proves that Generative models contain Discriminative information implicitly.
```
