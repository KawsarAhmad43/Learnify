# Real World Use Case: Drug Discovery

**Problem**: We want to find a new molecule that kills bacteria.
**Method**:
1.  Train a VAE on millions of known chemical strings (SMILES format).
2.  The VAE learns a continuous "Chemical Space".
    *   Point A = Aspirin.
    *   Point B = Ibuprofen.
3.  **Interpolation**: If we walk from A to B in latent space, we find valid molecules that are "half-Aspirin, half-Ibuprofen".
4.  **Optimization**: Search for points in latent space that have high predicted solubility.
**Result**: Generating valid, novel chemical structures to test in the lab.
