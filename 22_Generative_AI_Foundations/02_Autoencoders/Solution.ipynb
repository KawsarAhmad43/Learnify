# Solution: Compression Math

**Objective**: Data Size.

```python
original_bytes = 784
latent_bytes = 128

ratio = original_bytes / latent_bytes
print(f"Compression Ratio: {ratio:.2f}x")

# 2-bit quantization
quant_bytes = 32 * (2/8) # 2 bits is 0.25 bytes
ratio_quant = original_bytes / quant_bytes
print(f"Quantized Ratio: {ratio_quant:.2f}x")
```
