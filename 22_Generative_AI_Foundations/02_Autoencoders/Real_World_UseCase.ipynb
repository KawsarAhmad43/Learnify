# Real World Use Case: Industrial Defect Detection

**Scenario**: Manufacturing screws.
**Data**: Millions of photos of "Good" screws. almost zero photos of "Bad" screws (defects are rare).
**Method**:
1.  Train an Autoencoder ONLY on good screws.
2.  The Autoencoder becomes an expert at compressing/decompressing good screws.
3.  **Runtime**: Pass every screw on the conveyor belt through the AE.
    *   Good Screw -> Reconstruction Error 0.01.
    *   Bent Screw -> Reconstruction Error 0.5 (Model panics because it doesn't know how to compress 'bent').
4.  **Action**: If Error > 0.1, eject part.
**Result**: Automated QA without needing labeled defect data.

```python
reconstruction_error = 0.45
threshold = 0.1

if reconstruction_error > threshold:
    print("Anomaly Detected! Stops Machine.")
```
