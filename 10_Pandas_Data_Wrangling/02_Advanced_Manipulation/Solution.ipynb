# Solution: Advanced Manipulation

```python
import pandas as pd
import numpy as np

employees = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Department': ['HR', 'IT', 'IT', 'HR', 'Sales']
})

salaries = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4, 5, 1, 2],
    'Year': [2021, 2021, 2021, 2021, 2021, 2022, 2022],
    'Salary': [50000, 80000, 82000, 52000, 90000, 52000, 85000]
})

# 1. Merge
df = pd.merge(employees, salaries, on='EmployeeID')
print("Merged DF:")
display(df)

# 2. Avg Salary per Dept
avg_sal = df.groupby('Department')['Salary'].mean()
print("\nAvg Salary per Dept:")
display(avg_sal)

# 3. Highest Paid per Dept (using loc + idxmax or sort_values)
# Let's just consider the max salary record irrespective of year for simplicity
idx = df.groupby('Department')['Salary'].idxmax()
highest_paid = df.loc[idx]
print("\nHighest Paid Records per Dept:")
display(highest_paid)
```

### Task 2: Pivot Table

```python
pivot = df.pivot_table(index='Department', columns='Year', values='Salary', aggfunc='sum')
print("\nSalary Pivot Table:")
display(pivot)
```
