# Solution: Pandas Foundations

```python
import pandas as pd
import numpy as np
```

### Task 1: Basic Stats

```python
data = {
    'Product': ['Laptop', 'Mouse', 'Monitor', 'Keyboard', 'Phone', 'Tablet', 'Charger', 'Headphones', 'Webcam', 'Stand'],
    'Price': [1200, 25, 300, 50, 800, 400, 20, 150, 80, 40],
    'Category': ['Electronics', 'Accessories', 'Electronics', 'Accessories', 'Electronics', 'Electronics', 'Accessories', 'Audio', 'Electronics', 'Furniture'],
    'Rating': [4.5, 4.0, 4.2, 3.8, 4.7, 4.3, 2.5, 4.6, 3.5, 4.1]
}
df = pd.DataFrame(data)

# 1. Head
print("First 5 rows:")
display(df.head())

# 2. Describe
print("\nSummary Stats:")
display(df.describe())

# 3. Dtypes
print("\nData Types:")
print(df.dtypes)
```

### Task 2: Advanced Filtering

```python
# 1. Electronics > 500
premium_electronics = df[(df['Category'] == 'Electronics') & (df['Price'] > 500)]
print("\nPremium Electronics:")
display(premium_electronics)

# 2. Low Rating
low_rated = df.loc[df['Rating'] < 3.0, ['Product', 'Rating']]
print("\nLow Rated Items:")
display(low_rated)
```

### Task 3: Indexing

```python
# 1. Set Index
df_indexed = df.set_index('Product')

# 2. Select by Name
phone_row = df_indexed.loc['Phone']
print("\nPhone Row utilizing index:")
print(phone_row)
```
