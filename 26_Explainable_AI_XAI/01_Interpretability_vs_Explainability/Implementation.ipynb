# Implementation: Visualizing Coefficients

**Goal**: Interpret a Linear Model.

```python
import numpy as np
import pandas as pd

# 1. Mock Linear Regression Weights
features = ["Income", "Debt", "Age", "Years_Employed"]
weights = [0.8, -0.5, 0.1, 0.2]

# 2. Interpretation
df = pd.DataFrame({"Feature": features, "Weight": weights})
df = df.sort_values("Weight", ascending=False)

print("Model Logic:")
for index, row in df.iterrows():
    direction = "Increases" if row['Weight'] > 0 else "Decreases"
    print(f"{row['Feature']} {direction} the score by {abs(row['Weight'])} per unit.")
```
