# Real World Use Case: Credit Scoring (FICO)

**Scenario**: Bank denies a loan.
**Law**: In the US (ECOA) and EU (GDPR), the user has a "Right to Explanation". The bank MUST say "You were denied because your Debt-to-Income ratio is too high."
**Problem**: A Neural Network might deny them because of a complex interaction between "Zip Code" and "Shopping History" (which correlates with Race).
**Result**: Banks refuse to use deep learning for core credit decisions to avoid breaking the law. They use "Monotonic Gradient Boosted Trees" or Regression.
