# Implementation: Error Analysis Dashboard

**Goal**: Find the weak link.

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score

# 1. Mock Data
data = pd.DataFrame({
    "Outcome": [1, 1, 1, 0, 0, 0, 1, 1, 0, 0],
    "Prediction": [1, 1, 0, 0, 0, 1, 1, 1, 0, 0], # Errors at idx 2 and 5
    "Lighting": ["Day", "Day", "Night", "Day", "Day", "Night", "Day", "Day", "Night", "Night"]
})

# 2. Calculate Overall Accuracy
acc = accuracy_score(data["Outcome"], data["Prediction"])
print(f"Overall Acc: {acc:.2f}")

# 3. Slice Logic
for slice_name in data["Lighting"].unique():
    subset = data[data["Lighting"] == slice_name]
    slice_acc = accuracy_score(subset["Outcome"], subset["Prediction"])
    print(f"Slice '{slice_name}': {slice_acc:.2f} (Count: {len(subset)})")
```

## Conclusion
The model fails 50% of the time at Night. We need to collect more Night data.
