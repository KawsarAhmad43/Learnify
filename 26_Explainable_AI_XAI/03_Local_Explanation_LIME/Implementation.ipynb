# Implementation: LIME Logic (Simplified)

**Goal**: Local Linear Approximation.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Black Box (e.g., y = x^2)
# Non-linear function
def black_box(x):
    return x**2

# 2. Point to Explain (x = 3, y = 9)
x_target = 3.0

# 3. Perturb neighborhood
samples = np.random.normal(x_target, 0.1, 100)
predictions = black_box(samples)

# 4. Fit Linear Model
explainer = LinearRegression()
explainer.fit(samples.reshape(-1, 1), predictions)

print(f"Target Point: x={x_target}")
print(f"Correct Tangent Slope (Derivative 2x): {2 * x_target}")
print(f"LIME Slope: {explainer.coef_[0]:.2f}")
```

## Conclusion
LIME successfully recovered the derivative (slope) at that point.
