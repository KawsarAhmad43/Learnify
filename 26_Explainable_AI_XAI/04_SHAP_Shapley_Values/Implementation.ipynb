# Implementation: Analytical Shapley

**Goal**: Fair Division.

```python
# Simple Scenario: 2 Players (F1, F2)
# v(S): Value of subset S
# v({}) = 0
# v({F1}) = 10
# v({F2}) = 20
# v({F1, F2}) = 90 (Synergy!)

# Shapley Calculation for F1:
# 1. Order (F1, F2): F1 contributes v({F1}) - v({}) = 10 - 0 = 10
# 2. Order (F2, F1): F1 contributes v({F1, F2}) - v({F2}) = 90 - 20 = 70
# Average: (10 + 70) / 2 = 40

# Calculation for F2:
# 1. Order (F2, F1): F2 contributes 20 - 0 = 20
# 2. Order (F1, F2): F2 contributes 90 - 10 = 80
# Average: (20 + 80) / 2 = 50

# Check Additivity
# Total Payout = 40 (F1) + 50 (F2) = 90. Matches v({F1, F2}).

print("F1 Share: 40")
print("F2 Share: 50")
print("Even though F1 is weak alone (10), it is very valuable in the team (synergy).")
```

SHAP captures interactions accurately.
