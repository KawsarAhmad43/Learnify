# Implementation: Saliency Map Logic

**Goal**: Backprop to Pixels.

```python
import torch

# 1. Mock Image (3x3)
img = torch.tensor([[1.0, 2.0, 1.0], 
                    [0.0, 5.0, 0.0], 
                    [1.0, 2.0, 1.0]], requires_grad=True)

# 2. Mock CNN
def cnn(x):
    # It heavily relies on the center pixel
    return x[1, 1] * 10 + x[0, 0] * 2

# 3. Forward
score = cnn(img)
print(f"Output Score: {score.item()}")

# 4. Backward
score.backward()

# 5. Gradient (Saliency)
print("Saliency Map (Gradients):")
print(img.grad)
print("Notice the center pixel has grad=10. It is the most important.")
```

## Conclusion
In PyTorch, XAI is often just `.backward()`.
