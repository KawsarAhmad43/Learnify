# Implementation: Concept Vector Math

**Goal**: Linear Probing.

```python
import numpy as np

# 1. Mock Latent Space (2D for simplicity)
# Assume the model has learned a 2D representation of images

# 2. Concept: "Redness"
# We gather red images and random images.
# We find that 'Redness' aligns with the X-axis.
concept_vector = np.array([1.0, 0.0])

# 3. Image Embedding (A specific Apple)
img_embedding = np.array([0.5, 0.5])

# 4. Model Gradient at this embedding
# d(AppleScore) / d(Embedding)
# Assume the model REALLy likes X-axis features for Apples.
gradient = np.array([0.9, 0.1])

# 5. TCAV Score (Directional Derivative)
# Dot product of Gradient and Concept Vector
tcav_score = np.dot(gradient, concept_vector)

print(f"TCAV Score for Redness: {tcav_score:.2f}")
print("Positive score means Redness positively influences the Apple prediction.")
```
