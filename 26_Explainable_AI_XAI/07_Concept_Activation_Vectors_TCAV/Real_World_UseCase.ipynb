# Real World Use Case: Bias Detection in Medical AI

**Scenario**: Skin Cancer Detector.
**Hypothesis**: Is the model biasing towards "Male" skin? (Maybe training data had more men).
**TCAV**: 
1.  Define "Male" concept (using photos of men).
2.  Test the "Cancer" class sensitivity to the "Male" concept.
**Result**: TCAV Score is 0.0 (Neutral). The model does NOT use gender to predict cancer.
**Result 2**: Test "Ruler" concept (dermatologists often put rulers next to tumors). TCAV Score is 0.8 (High). 
**Alert**: The model is detecting the Ruler, not the Tumor!
