# Implementation: Nearest Neighbor Counterfactual

**Goal**: Find the closest point on the other side.

```python
import numpy as np

# 1. Mock Data (People who got approved)
approved_data = np.array([
    [50000, 700], # Income, Score
    [80000, 650],
    [45000, 720]
])

# 2. Rejected User
user = np.array([40000, 700])

# 3. Find Closest Approved Person (L1 Distance)
distances = np.abs(approved_data - user).sum(axis=1)
closest_idx = np.argmin(distances)
cf = approved_data[closest_idx]

print(f"User: {user}")
print(f"Closest Approved Profile: {cf}")
print(f"Advice: Increase Income by {cf[0] - user[0]} or Change Score to {cf[1]}.")
print("Note: This is a simplistic KNN approach. Real methods use optimization.")
```

## Conclusion
This gives the user a 'Target' to aim for.
