# Implementation: Simple Additive Masking

**Goal**: Cryptographic Privacy.

```python
import random

# 1. Two Users, One Secret
value_a = 10
value_b = 20
secret = random.randint(0, 1000)
print(f"Secret Mask: {secret}")

# 2. Encryption
msg_a = value_a + secret
msg_b = value_b - secret

# 3. Transmission (Server sees meaningless numbers)
print(f"Server sees A: {msg_a}")
print(f"Server sees B: {msg_b}")

# 4. Aggregation
total = msg_a + msg_b
print(f"True Sum: {total}")
```

## Conclusion
The server learned the Sum (30) but does not know if A contributed 0 and B contributed 30, or vice versa.
