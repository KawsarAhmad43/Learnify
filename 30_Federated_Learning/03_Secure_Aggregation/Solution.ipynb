# Solution: Median Aggregation

**Objective**: Robustness.

```python
import torch

updates = torch.tensor([1.0, 1.1, 1.0, 1000.0, 0.9]) # 1000.0 is an attack

mean_agg = updates.mean()
median_agg = updates.median()

print(f"Mean (Poisoned): {mean_agg:.2f}")
print(f"Median (Robust): {median_agg:.2f}")
```
