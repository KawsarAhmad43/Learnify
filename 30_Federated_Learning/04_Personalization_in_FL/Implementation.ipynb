# Implementation: Fine-Tuning Strategy

**Goal**: Adaptation.

```python
# 1. Global Model arrives
global_weights = 10.0

# 2. User A's Data (Requires lower weight)
user_a_target = 8.0
user_a_model = global_weights

# 3. Local Fine-Tuning (Gradient Descent)
lr = 0.1
for epoch in range(5):
    loss = (user_a_model - user_a_target)**2
    grad = 2 * (user_a_model - user_a_target)
    user_a_model -= lr * grad
    
print(f"Global Logic: {global_weights}")
print(f"Personalized Logic: {user_a_model:.2f}")
print("The user enjoys a model tailored to them, but initialized by the world.")
```
