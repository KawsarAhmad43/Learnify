# Real World Use Case: Predictive Text (Gboard)

**Scenario**: Google wants to predict the next word you type.
**Privacy**: Google CANNOT upload your keystrokes to the cloud. That's a massive violation.
**Solution (FL)**:
1.  Your phone downloads a generic English model.
2.  You type "See you at the [Beach]".
3.  The model predicted "[Office]". Error!
4.  Your phone updates the model LOCALLY to predict "Beach".
5.  At night, your phone sends the *weights update* to Google.
**Result**: Google's global model learns "People go to the beach on weekends" without knowing *who* went to the beach.
