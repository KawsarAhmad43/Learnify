# Implementation: FedAvg Simulation

**Goal**: Aggregate 3 local models.

```python
import torch

# 1. Global Model (Simple Linear Layer)
global_model = torch.tensor([1.0, 1.0])

# 2. Client Updates (Simulated)
# Client A saw data that pushed weights up
w_a = torch.tensor([1.2, 1.1]) 
# Client B saw data that pushed weights down
w_b = torch.tensor([0.8, 0.9])
# Client C saw neutral data
w_c = torch.tensor([1.0, 1.0])

# 3. Server Aggregation (FedAvg)
new_global_model = (w_a + w_b + w_c) / 3

print(f"Old Global: {global_model}")
print(f"New Global: {new_global_model}")
print("The Global Model learned from A and B without seeing their data.")
```

## Conclusion
It's just averaging. The magic is in the distributed system.
