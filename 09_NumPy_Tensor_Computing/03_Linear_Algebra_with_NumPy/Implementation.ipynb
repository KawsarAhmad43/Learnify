# Implementation: Linear Algebra

We will simulate the forward pass of a single Neural Network layer.

```python
import numpy as np
```

## 1. Neural Network Layer
$$Output = Activation(X \cdot W + b)$$
Where $X$ is input, $W$ is weights, $b$ is bias.

```python
# Inputs: Batch of 3 samples, 4 features each
X = np.random.randn(3, 4)

# Weights: 4 features, 5 neurons (outputs)
W = np.random.randn(4, 5)

# Bias: 5 neurons
b = np.random.randn(5)

# Linear Step: Z = X @ W + b
Z = X @ W + b

print("Input Shape:", X.shape)
print("Weight Shape:", W.shape)
print("Output Shape:", Z.shape)
print("Output (Z):\n", Z)
```

## 2. Activation (ReLU)

```python
A = np.maximum(0, Z)
print("Activated Output:\n", A)
```
