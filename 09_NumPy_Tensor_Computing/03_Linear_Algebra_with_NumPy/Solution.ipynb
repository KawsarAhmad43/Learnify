# Solution: Linear Algebra

```python
import numpy as np
```

### Task 1: System of Equations

```python
A = np.array([[3, 1], [1, 2]])
b = np.array([9, 8])

x = np.linalg.solve(A, b)
print("Solution [x, y]:", x)

# Verify
print("Verification (A @ x):", A @ x)
```

### Task 2: Cosine Similarity

```python
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

dot_product = np.dot(u, v)
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)

sim = dot_product / (norm_u * norm_v)
print(f"Cosine Similarity: {sim:.4f}")
```
