# Solution: NumPy Foundations

```python
import numpy as np
```

### Task 1: Create and Reshape

```python
arr = np.arange(24)
arr_3d = arr.reshape(2, 3, 4)
print("Shape:", arr_3d.shape)
print("Dtype:", arr_3d.dtype)
print(arr_3d)
```

### Task 2: Image Slicing

```python
np.random.seed(42)
image = np.random.randint(0, 256, (10, 10))
print("Original Image (Top Left):")
print(image[:5, :5])

# Center crop rows 3:7, cols 3:7 (indices 3,4,5,6)
image[3:7, 3:7] = 0

print("\nMasked Image Center:")
print(image)
```

### Task 3: Conditionals

```python
ints = np.random.randint(0, 100, 20)
print("Original:", ints)

# Replace odd numbers (where x % 2 == 1)
ints[ints % 2 == 1] = -1
print("Modified:", ints)
```
