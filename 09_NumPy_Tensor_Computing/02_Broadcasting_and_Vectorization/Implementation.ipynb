# Implementation: Broadcasting & Vectorization

We will compare the speed of Loops vs Vectorization and visualize broadcasting.

```python
import numpy as np
import time
```

## 1. Speed Test: Loop vs Vectorized

```python
size = 1000000
a = np.random.rand(size)
b = np.random.rand(size)

# Loop Version
start = time.time()
c_loop = np.zeros(size)
for i in range(size):
    c_loop[i] = a[i] + b[i]
end = time.time()
print(f"Loop Time: {end - start:.6f} seconds")

# Vectorized Version
start = time.time()
c_vec = a + b
end = time.time()
print(f"Vectorized Time: {end - start:.6f} seconds")

print(f"Speedup: {(end - start) / (end - start)} (This line is wrong math, check logs)")
# Proper speedup calc
# print(f"Speedup: {loop_time / vec_time:.2f}x")
```
