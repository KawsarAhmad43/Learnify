# Regression Metrics Solutions

## Task 1: RMSE

```python
import numpy as np
from sklearn.metrics import mean_squared_error

y_true = np.array([100, 200, 300, 400, 500])
y_pred_A = np.array([110, 190, 310, 390, 510])
y_pred_B = np.array([100, 200, 300, 400, 1000])

rmse_A = np.sqrt(mean_squared_error(y_true, y_pred_A))
rmse_B = np.sqrt(mean_squared_error(y_true, y_pred_B))

print(f"RMSE A: {rmse_A:.2f}")
print(f"RMSE B: {rmse_B:.2f}")
```

## Task 2: Baseline

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

data = fetch_california_housing()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)

# Dummy Model (Median)
median_price = np.median(y_train)
y_pred_dummy = np.full(shape=y_test.shape, fill_value=median_price)

mae = mean_absolute_error(y_test, y_pred_dummy)
print(f"Baseline MAE: {mae:.4f} (Hundreds of thousands of dollars)")
```
