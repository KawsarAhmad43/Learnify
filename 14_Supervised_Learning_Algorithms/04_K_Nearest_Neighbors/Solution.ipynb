# Solution: Fruit Evaluation

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# 1. Data
X = np.array([[150, 7.5], [170, 8.0], [80, 6.0], [75, 5.8], [200, 8.5]])
y = ['Apple', 'Apple', 'Lemon', 'Lemon', 'Apple']

# 2. Train
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)

# 3. Predict
unknown = [[160, 7.0]]
print(f"Prediction for {unknown}: {knn.predict(unknown)[0]}")
```
