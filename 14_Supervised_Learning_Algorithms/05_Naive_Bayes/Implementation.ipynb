# Implementation: Naive Bayes (Spam Filtering)

Classifying text messages.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

# 1. Data
emails = [
    "Win money now free cash",
    "Meeting at noon please be late",
    "Free viagra now",
    "Submit your report by Friday",
    "Congratulations you won a lottery"
]
labels = ['spam', 'ham', 'spam', 'ham', 'spam']

# 2. Pipeline (Vectorize -> NB)
model = make_pipeline(CountVectorizer(), MultinomialNB())
model.fit(emails, labels)

# 3. Predict
new_email = ["free cash won"]
print(f"Prediction for '{new_email[0]}': {model.predict(new_email)[0]}")
```
