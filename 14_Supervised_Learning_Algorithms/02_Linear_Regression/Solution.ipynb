# Solution: Salary Prediction

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Data
experience = np.array([1.1, 1.3, 1.5, 2.0, 2.2, 2.9, 3.0, 3.2, 3.2, 3.7]).reshape(-1, 1)
salary = np.array([39343, 46205, 37731, 43525, 39891, 56642, 60150, 54445, 64445, 57189])

# 2. Plot
plt.scatter(experience, salary, color='blue')
plt.title('Salary vs Experience')
plt.xlabel('Years')
plt.ylabel('Salary')
plt.show()

# 3. Train
model = LinearRegression()
model.fit(experience, salary)

# 4. Predict
prediction = model.predict([[5.5]])
print(f"Predicted salary for 5.5 years: ${prediction[0]:,.2f}")

# Visualize Line
plt.scatter(experience, salary, color='blue')
plt.plot(experience, model.predict(experience), color='red')
plt.show()
```
