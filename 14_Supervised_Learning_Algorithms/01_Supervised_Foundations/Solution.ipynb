# Solution: Diamonds Task

```python
import seaborn as sns
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score

# Load Data
df = sns.load_dataset('diamonds').sample(1000, random_state=42) # Downsample for speed

X = df[['carat']]

# 1. Regression
y_reg = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y_reg, test_size=0.2, random_state=42)

reg = LinearRegression()
reg.fit(X_train, y_train)
print(f"Regression MSE: {mean_squared_error(y_test, reg.predict(X_test)):.2f}")

# 2. Classification
df['is_expensive'] = (df['price'] > 5000).astype(int)
y_class = df['is_expensive']
X_train, X_test, y_train, y_test = train_test_split(X, y_class, test_size=0.2, random_state=42)

clf = LogisticRegression()
clf.fit(X_train, y_train)
print(f"Classification Accuracy: {accuracy_score(y_test, clf.predict(X_test)):.2f}")
```
