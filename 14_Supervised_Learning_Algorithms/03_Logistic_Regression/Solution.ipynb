# Solution: Purchase Prediction

```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# 1. Data
age = np.array([22, 25, 47, 52, 46, 56, 55, 60, 62, 63]).reshape(-1, 1)
bought = np.array([0, 0, 1, 0, 1, 1, 1, 1, 1, 1])

# 2. Train
model = LogisticRegression()
model.fit(age, bought)

# 3. Probability for 40
prob_40 = model.predict_proba([[40]])[0][1]
print(f"Probability a 40-year-old buys: {prob_40:.2%}")

# 4. Prediction for 30
pred_30 = model.predict([[30]])
print(f"Will a 30-year-old buy? {'Yes' if pred_30[0]==1 else 'No'}")
```
