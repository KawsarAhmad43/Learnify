# Implementation: Stock Price Prediction

**Goal**: Sequence Prediction.

```python
import torch
import torch.nn as nn

# 1. Model
class LSTMForecaster(nn.Module):
    def __init__(self):
        super().__init__()
        self.lstm = nn.LSTM(input_size=1, hidden_size=32, batch_first=True)
        self.fc = nn.Linear(32, 1)
        
    def forward(self, x):
        # x shape: [batch, seq_len, features]
        out, _ = self.lstm(x)
        # Take the last time step
        last_hidden = out[:, -1, :]
        return self.fc(last_hidden)

# 2. Data (Window of 5 days)
input_seq = torch.tensor([[[10.0], [11.0], [12.0], [13.0], [14.0]]]) # Shape (1, 5, 1)
target = torch.tensor([[15.0]])

# 3. Train
model = LSTMForecaster()
pred = model(input_seq)
print(f"Prediction: {pred.item():.2f}")
print("The LSTM processes the sequence [10, 11, 12, 13, 14] and predicts the continuation.")
```

## Conclusion
Note: Stock prediction is hard. This model will likely just predict 'Yesterday's Price' (Persistence Model).
