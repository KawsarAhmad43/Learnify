# Implementation: Testing for Stationarity (ADF Test)

**Goal**: Statistics.

```python
import numpy as np
from statsmodels.tsa.stattools import adfuller

# 1. Create Non-Stationary Data (Trend)
time = np.arange(100)
trend = time * 0.5
noise = np.random.normal(0, 5, 100)
data = trend + noise

# 2. Augmented Dickey-Fuller Test
result = adfuller(data)
p_value = result[1]

print(f"P-Value: {p_value:.4f}")
if p_value > 0.05:
    print("Fail to reject Null Hypothesis -> Data is NON-Stationary.")
else:
    print("Data is Stationary.")

# 3. Make it Stationary (Differencing)
diff_data = np.diff(data)
p_value_diff = adfuller(diff_data)[1]
print(f"Differenced P-Value: {p_value_diff:.4f} (Stationary now!)")
```

## Conclusion
Always difference your financial data before feeding it to a model.
