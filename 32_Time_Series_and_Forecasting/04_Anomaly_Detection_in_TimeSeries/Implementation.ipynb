# Implementation: Isolation Forest

**Goal**: Outlier Detection.

```python
from sklearn.ensemble import IsolationForest
import numpy as np

# 1. Data
X = np.array([[-1.1], [0.3], [0.5], [100.0]]) # 100.0 is outlier

# 2. Model
clf = IsolationForest(random_state=0).fit(X)

# 3. Predict
preds = clf.predict(X)
print(f"Predictions: {preds}")
print("1 = Inlier, -1 = Outlier")
print("Notice 100.0 is flagged as -1.")
```

## Conclusion
Simple but effective for tabular anomalies.
