# Linear Algebra Implementation with NumPy

## 1. Vectors and Operations

```python
import numpy as np

# Defining vectors
v = np.array([1, 2, 3])
print(f"Vector v: {v}")
print(f"Shape: {v.shape}")

# Scalar Multiplication
print(f"2 * v: {2 * v}")

# Element-wise functions
print(f"v squared: {v**2}")
```

## 2. Matrices and Dimensions

```python
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print(f"Matrix A:\n{A}")
print(f"Shape of A: {A.shape}")

# Transpose (rows become columns)
print(f"Transpose of A:\n{A.T}")
```

## 3. Dot Product

```python
a = np.array([1, 2])
b = np.array([3, 4])

dot_prod = np.dot(a, b)
# Calculation: 1*3 + 2*4 = 3 + 8 = 11
print(f"Dot product: {dot_prod}")

# Alternative syntax (@ operator)
print(f"Using @: {a @ b}")
```
