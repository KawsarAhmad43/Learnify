# Linear Algebra Tasks: Solutions

```python
import numpy as np
```

## Task 1: NumPy Basics

```python
v1 = np.array([2, 3])
v2 = np.array([4, -1])

# 1. Dot Product
dot_manual = (2*4) + (3*-1)
dot_np = np.dot(v1, v2)
print(f"Manual: {dot_manual}, NumPy: {dot_np}")

# 2. L2 Norm (Euclidean Distance from origin)
# sqrt(2^2 + 3^2) = sqrt(4+9) = sqrt(13)
l2_norm = np.linalg.norm(v1)
print(f"L2 Norm of v1: {l2_norm:.4f}")
```

## Task 2: Matrix Multiplication

```python
A = np.random.randint(1, 10, (2, 3))
B = np.random.randint(1, 10, (3, 2))

print(f"A shape: {A.shape}")
print(f"B shape: {B.shape}")

C = np.dot(A, B) # Or A @ B
print(f"C (A @ B) shape: {C.shape}") # Should be (2, 2)
print(C)

D = np.dot(B, A)
print(f"D (B @ A) shape: {D.shape}") # Should be (3, 3)
print(D)
```

## Task 3: Cosine Similarity

```python
def cosine_similarity(v1, v2):
    dot = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    
    if norm_v1 == 0 or norm_v2 == 0:
        return 0
        
    return dot / (norm_v1 * norm_v2)

vec_a = np.array([1, 0])
vec_b = np.array([0, 1])
print(f"Sim([1,0], [0,1]): {cosine_similarity(vec_a, vec_b)}")

vec_c = np.array([1, 1])
vec_d = np.array([2, 2])
print(f"Sim([1,1], [2,2]): {cosine_similarity(vec_c, vec_d)}")
```
