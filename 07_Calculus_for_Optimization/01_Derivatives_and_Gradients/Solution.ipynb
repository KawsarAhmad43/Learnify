# Solution: Derivatives & Gradients

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

### Task 1: Partial Derivatives

```python
x, y = sp.symbols('x y')
f = 3*x**2 + 2*y**3 - 5*x*y

# Partial wrt x
df_dx = sp.diff(f, x)
print(f"df/dx: {df_dx}")

# Partial wrt y
df_dy = sp.diff(f, y)
print(f"df/dy: {df_dy}")

# Evaluate at (1, 2)
val_x = df_dx.evalf(subs={x: 1, y: 2})
val_y = df_dy.evalf(subs={x: 1, y: 2})

print(f"Gradient at (1, 2): [{val_x}, {val_y}]")
```

### Task 2: Gradient Vector Visualization

```python
x_vals = np.linspace(-np.pi, np.pi, 100)
y_vals = np.linspace(-np.pi, np.pi, 100)
X, Y = np.meshgrid(x_vals, y_vals)
Z = np.sin(X) * np.cos(Y)

plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Z, cmap='viridis', levels=20)
plt.colorbar(contour)
plt.title('Contour plot of sin(x)cos(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
