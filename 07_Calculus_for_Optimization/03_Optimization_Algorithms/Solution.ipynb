# Solution: Optimization Algorithms

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Task 1: 2D Gradient Descent

```python
def df_dx(x):
    return 2*x

def df_dy(y):
    return 2*y

def gradient_descent_2d(start_x, start_y, lr, epochs):
    x = start_x
    y = start_y
    path = []
    
    for _ in range(epochs):
        grad_x = df_dx(x)
        grad_y = df_dy(y)
        
        x = x - lr * grad_x
        y = y - lr * grad_y
        path.append((x, y))
        
    return x, y, path

final_x, final_y, path = gradient_descent_2d(3, 4, 0.1, 50)
print(f"Final Position: ({final_x:.4f}, {final_y:.4f})")
print("It converged to (0,0).")
```

### Task 2: Effect of Learning Rate

```python
# Large LR (Overshooting)
_, _, path_large = gradient_descent_2d(3, 4, 1.1, 10)
print(f"Path with LR=1.1: {path_large[:3]} ... (Exploding)")

# Small LR (Slow)
_, _, path_small = gradient_descent_2d(3, 4, 0.0001, 10)
print(f"Path with LR=0.0001: {path_small[:3]} ... (Barely moving)")
```
