# Real World Use Case: Model Registry (MLflow)

**Scenario**: You trained 50 models today. Which one was the best?
**Problem**: "I think it was `model_final_v2_fixed.pkl` on my Desktop".
**Solution**: MLflow Tracking Server.
**Workflow**:
```python
with mlflow.start_run():
    mlflow.log_param("lr", 0.01)
    model.fit(X, y)
    mlflow.log_metric("rmse", 12.5)
    mlflow.sklearn.log_model(model, "model")
```
**Result**: A searchable database of every experiment you've ever run, with the exact code commit hash and hyperparameters.
