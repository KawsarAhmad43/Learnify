# Implementation: PyTorch Geometric Basics

**Goal**: Represent a graph.

```python
import torch
# Note: In real life you'd pip install torch_geometric

# 1. Nodes (3 people, 2 features each: [Age, Income])
x = torch.tensor([
    [25, 50], # Node 0
    [30, 80], # Node 1
    [20, 10]  # Node 2
], dtype=torch.float)

# 2. Edges (Connectivity)
# 0->1, 1->2
edge_index = torch.tensor([
    [0, 1], # Source
    [1, 2]  # Target
], dtype=torch.long)

print(f"Node Matrix Shape: {x.shape}")
print(f"Edge List Shape: {edge_index.shape}")
print("This 'Sparse' representation is much better than a dense Adjacency Matrix for large graphs.")
```

## Conclusion
Graph data = Features + Topology.
