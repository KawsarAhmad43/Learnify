# Implementation: Time Series & Image Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits

sns.set_theme()
```

## 1. Time Series (Stock Simulation)

```python
dates = pd.date_range('2023-01-01', periods=100)
stock_a = np.random.randn(100).cumsum() + 100
stock_b = np.random.randn(100).cumsum() + 100

plt.figure(figsize=(10, 5))
plt.plot(dates, stock_a, label='Stock A')
plt.plot(dates, stock_b, label='Stock B')
plt.title('Stock Price Comparison')
plt.legend()
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

## 2. Image Grid (MNIST Digits)

```python
digits = load_digits()
images = digits.images[:16]

fig, axes = plt.subplots(4, 4, figsize=(6, 6))
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap='binary')
    ax.axis('off')
    
plt.suptitle('MNIST Digits Sample')
plt.show()
```
