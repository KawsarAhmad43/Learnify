# Solution: Data Embedding

**Objective**: Input.
1. **Basis Encoding**: $x=5 (101)$ -> $|101\rangle$. (Efficient but limited arithmetic).
2. **Amplitude Encoding**: $x=[0.6, 0.8]$ -> $0.6|0\rangle + 0.8|1\rangle$. (Compact, stores measurements in amplitudes, hard to prepare).
3. **Angle Encoding**: $x=\pi$ -> $R_x(\pi)|0\rangle$. (Standard for NISQ).
