# Solution: Parameter Shift Rule

**Objective**: Gradients.
1. **Problem**: You cannot do Backprop through a Quantum Measurement (It's not differentiable, it's stochastic).
2. **Solution**: For gates like $R_x(\theta)$, the analytic gradient is exactly:
   $\nabla \theta = 0.5 * (f(\theta + \pi/2) - f(\theta - \pi/2))$.
3. **Conclusion**: To get the gradient for 1 parameter, we must run the circuit TWICE (shifted forward and backward). If you have 1000 parameters, that's 2000 runs per step.
