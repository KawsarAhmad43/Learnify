# Real World Use Case: Gate Assignment at Airports

**Scenario**: Heathrow Airport.
**Problem**: 100 Planes land. 50 Gates available.
**Constraints**:
*   Big plane needs big gate.
*   Connecting flights should be close.
*   Minimize passenger walking time.
**Solution**: This is a constrained optimization problem (QUBO). It can be mapped to an Ising Model and solved with QAOA.
**Status**: Currently done with Classical Heuristics. Quantum promises a "Better" global optimum (fewer delays).
