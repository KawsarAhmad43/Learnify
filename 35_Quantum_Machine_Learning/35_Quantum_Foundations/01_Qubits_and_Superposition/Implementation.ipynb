# Implementation: Creating a Qubit in PennyLane

**Goal**: Hello Quantum World.

```python
import pennylane as qml
from pennylane import numpy as np

# 1. Define a Device (Simulator)
dev = qml.device("default.qubit", wires=1)

# 2. Define a Quantum Node (Circuit)
@qml.qnode(dev)
def circuit():
    # Start in |0>
    # Apply Hadamard Gate (Creates Superposition)
    qml.Hadamard(wires=0)
    # Measure probabilities
    return qml.probs(wires=0)

# 3. Run
print("Probabilities [0, 1]:", circuit())
print("You should see [0.5, 0.5]. Ideally, it is 50% 0 and 50% 1.")
```

## Conclusion
You just simulated a perfect coin flip using Quantum Mechanics.
