# Real World Use Case: Quantum Internet

**Scenario**: Unhackable Communication (QKD).
**Method**: E91 Protocol.
**Logic**:
1.  Alice and Bob share entangled pairs.
2.  They measure them to generate a secret key.
3.  If a hacker (Eve) tries to intercept the key, she must measure the qubits.
4.  Her measurement **disturbs** the entanglement (collapses the state).
5.  Alice and Bob detect the disturbance (Error rate goes up).
**Result**: They know instantly they are being watched and discard the key.
