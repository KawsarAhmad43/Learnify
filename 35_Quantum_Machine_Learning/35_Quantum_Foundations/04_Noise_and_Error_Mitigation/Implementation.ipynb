# Implementation: Simulating Noise

**Goal**: Reality Check.

```python
import pennylane as qml

# 1. Define a Noisy Device
# "default.mixed" handles mixed states (noisy density matrices)
dev = qml.device("default.mixed", wires=1)

@qml.qnode(dev)
def noisy_circuit():
    qml.Hadamard(wires=0)
    
    # 2. Add Bit Flip Noise (5% chance of flipping)
    qml.BitFlip(0.05, wires=0)
    
    return qml.probs(wires=0)

print("Noisy Probabilities:", noisy_circuit())
print("Instead of [0.5, 0.5], you might see [0.52, 0.48].")
```
