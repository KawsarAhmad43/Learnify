# Solution: SPSA (Simultaneous Perturbation Stochastic Approximation)

**Objective**: Optimization with few shots.
1. **Method**: Instead of tweaking parameters one by one, tweak ALL of them randomly at once (perturb by random vector $\Delta$).
2. **Gradient**: Estimate gradient as $(L(\theta + \Delta) - L(\theta - \Delta))$.
3. **Cost**: Only 2 runs regardless of the number of parameters.
4. **Result**: Noisy but effective for high-dimensional optimization on real hardware.
