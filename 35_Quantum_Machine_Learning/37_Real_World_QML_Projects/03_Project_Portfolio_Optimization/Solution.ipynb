# Solution: CVaR (Conditional Value at Risk)

**Objective**: Better Metrics.
1. **Mean**: Average return. (Good). 
2. **CVaR**: "Average of the worst 5% cases". (Better).
3. **Quantum CVaR**: Instead of optimizing the expectation value <H>, we optimize the "Tail probability" of the measurements. This focuses the optimizer on avoiding catastrophic losses.
