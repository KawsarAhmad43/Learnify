# Solution: Inferential Statistics

```python
import numpy as np
from scipy import stats
from statsmodels.stats.proportion import proportions_ztest
```

### Task 1: Store Conversion Rates
Using Z-test for proportions.

```python
count = np.array([120, 150])
nobs = np.array([1000, 1000])

stat, pval = proportions_ztest(count, nobs)
print(f"Z-statistic: {stat:.4f}")
print(f"P-value: {pval:.4f}")

if pval < 0.05:
    print("Conclusion: Page B is significantly better.")
else:
    print("Conclusion: No significant difference.")
```

### Task 2: Model Evaluation
Paired T-test for model errors.

```python
np.random.seed(123)
# Simulate errors (Model 2 is slightly better/lower error)
errors_model_1 = np.random.normal(loc=10, scale=2, size=50)
errors_model_2 = errors_model_1 - np.random.normal(loc=1, scale=0.5, size=50) # Systematically lower

# 1. Paired T-Test
t_stat, p_val = stats.ttest_rel(errors_model_1, errors_model_2)
print(f"Paired T-Test P-value: {p_val:.6f}")

# 2. Shapiro-Wilk Test for Normality
stat_1, p_shapiro_1 = stats.shapiro(errors_model_1)
print(f"Shapiro P-value (Model 1): {p_shapiro_1:.4f}")

if p_shapiro_1 > 0.05:
    print("Model 1 errors look Normal.")
else:
    print("Model 1 errors do NOT look Normal.")
```
