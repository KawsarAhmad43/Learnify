# Implementation: Mock Dockerfile

**Goal**: Infrastructure as Code.

```python
dockerfile_content = """
# 1. Base Image
FROM pytorch/pytorch:2.0.0-cuda11.7-cudnn8-runtime

# 2. Work Directory
WORKDIR /app

# 3. Dependencies
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt

# 4. Source Code
COPY src/ .

# 5. Entrypoint
EXPOSE 8000
CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"]
"""

print("Here is your Dockerfile:")
print(dockerfile_content)
print("Command to build: docker build -t my_model_app .")
print("Command to run: docker run -p 8000:8000 my_model_app")
```

## Conclusion
This file guarantees that if it works on your laptop, it WILL work on the server.
