# Real World Use Case: GitHub Copilot Deployment

**Scenario**: OpenAI updates the Codex model.
**Pipeline**:
1.  **Safety Check**: Run thousands of prompts to check if the new model outputs hate speech.
2.  **Performance Check**: Check latency (did it get slower?).
3.  **Canary Deployment**: Roll out to 1% of users.
4.  **Monitor**: Watch for crash reports.
5.  **Full Rollout**: 100% of users.
