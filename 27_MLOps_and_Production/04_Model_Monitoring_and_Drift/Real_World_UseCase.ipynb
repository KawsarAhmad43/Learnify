# Real World Use Case: COVID-19 vs Instacart

**Scenario**: March 2020.
**Model**: Recommendation Engine ("People who buy chips buy soda").
**Drift**: Suddenly, everyone is buying Toilet Paper and Hand Sanitizer.
**Failure**: The model (trained on 2019 data) had no idea these items were correlated.
**Result**: Recommendations were nonsense.
**Fix**: Teams had to rapidly retrain models on *only* the last 2 weeks of data, discarding 2019 data as "obsolete".
