# Implementation: KS Test for Drift

**Goal**: Statistical Warning System.

```python
import numpy as np
from scipy.stats import ks_2samp

# 1. Reference Data (Training)
# Assume 'Income' feature
training_data = np.random.normal(50000, 10000, 1000)

# 2. Production Data (Today's traffic)
# Uh oh, inflation happened, mean shifted to 60k
prod_data = np.random.normal(60000, 10000, 1000)

# 3. Run Test
statistic, p_value = ks_2samp(training_data, prod_data)

print(f"KS Statistic: {statistic:.2f}")
print(f"P-Value: {p_value}")

if p_value < 0.05:
    print("DRIFT DETECTED! The distributions are significantly different.")
else:
    print("ALL GOOD. Distributions match.")
```
