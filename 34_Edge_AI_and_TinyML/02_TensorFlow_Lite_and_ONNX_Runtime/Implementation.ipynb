# Implementation: PyTorch to ONNX

**Goal**: Interoperability.

```python
import torch
import torch.nn as nn

# 1. Define Model
model = nn.Linear(10, 5)
dummy_input = torch.randn(1, 10)

# 2. Export
torch.onnx.export(
    model, 
    dummy_input, 
    "model.onnx", 
    input_names=["input"], 
    output_names=["output"]
)

print("Model saved as model.onnx.")
print("You can now load this in C++ using 'onnxruntime' library.")
```

## Conclusion
Write once, run everywhere.
