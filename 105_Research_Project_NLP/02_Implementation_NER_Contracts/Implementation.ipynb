## 1. Data Distribution
**Context**: Token imbalances standard in NER (Most tokens are "O" - Outside). 
If 'B-PER' is 0.01% of data, the model might optimize to predict 'O' everywhere.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Mock Tag Counts
tags = {'O': 50000, 'B-PER': 200, 'I-PER': 150, 'B-ORG': 300, 'I-ORG': 400}

plt.figure(figsize=(10, 5))
sns.barplot(x=list(tags.keys()), y=list(tags.values()))
plt.yscale('log') # Log scale because 'O' dominates
plt.title("Token Class Distribution (Log Scale)")
plt.show()
```

**Observation**: The 'O' class is 100x larger. We rely on the CrossEntropyLoss's ability to handle this naturally, but we must classify the minority classes correctly.
