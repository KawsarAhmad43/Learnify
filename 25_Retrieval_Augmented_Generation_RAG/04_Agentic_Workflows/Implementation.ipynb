# Implementation: Router Logic

**Goal**: Decide where to look.

```python
# Mock LLM Router
def router(query):
    # In reality, this is an LLM call: "Classify the intent of this query: [Greeting, Technical, Financial]"
    query = query.lower()
    if "hello" in query or "hi" in query:
        return "NO_RETRIEVAL"
    if "price" in query or "cost" in query:
        return "SQL_DB"
    return "VECTOR_DB"

queries = [
    "Hello there!",
    "How do I reset my password?",
    "What was the total cost of server maintenance?"
]

for q in queries:
    action = router(q)
    print(f"Query: '{q}' -> Action: {action}")
```

## Conclusion
Routing saves money (no embedding/search cost for greetings) and improves accuracy (sending math questions to SQL).
