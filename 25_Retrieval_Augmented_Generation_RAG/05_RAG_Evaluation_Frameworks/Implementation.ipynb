# Implementation: Faithfulness Metric

**Goal**: Automate grading.

```python
def calculate_faithfulness(context, answer):
    # In reality, this calls an LLM Judge
    context_facts = set(context.lower().split())
    answer_facts = set(answer.lower().split())
    
    # Naive overlap
    supported = 0
    total = len(answer_facts)
    
    for word in answer_facts:
        if word in context_facts:
            supported += 1
            
    return supported / total if total > 0 else 0

ctx = "Paris is the capital of France."
ans_good = "The capital is Paris."
ans_bad = "The capital is London."

print(f"Faithfulness (Good): {calculate_faithfulness(ctx, ans_good):.2f}")
print(f"Faithfulness (Bad): {calculate_faithfulness(ctx, ans_bad):.2f}")

print("LLM Judges are much smarter than this naive word overlap.")
```

## Conclusion
If Faithfulness < 0.8, your model is hallucinating. Reduce 'Temperature'.
