# Real World Use Case: CICD for Prompts

**Scenario**: You change the system prompt from "Be helpful" to "Be concise".
**Risk**: Did this break the accuracy?
**Pipeline**:
1.  **Test Set**: 100 Golden Q&A pairs.
2.  **Run**: Run the pipeline with the new prompt.
3.  **Eval**: Calculate Average RAGAS Score.
4.  **Gate**: If score drops > 2%, fail the deployment.
**Result**: Safe evolution of the chatbot.
