# Implementation: Mock Graph Traversal

**Goal**: Structured retrieval.

```python
# 1. Graph Data (Adjacency List)
graph = {
    "Elon Musk": [("CEO_of", "Tesla"), ("CEO_of", "SpaceX")],
    "Tesla": [("Competes_with", "BYD"), ("Located_in", "Texas")],
    "SpaceX": [("Located_in", "Texas")],
    "BYD": [("Located_in", "China")]
}

# 2. Query: "What competitors does Elon's car company have?"

# 3. Graph Logic (Multi-hop)
# Step A: Identify Entity "Elon Musk"
start_node = "Elon Musk"

# Step B: Traverse "CEO_of"
companies = [target for rel, target in graph[start_node] if rel == "CEO_of"]
# companies = [Tesla, SpaceX]

# Step C: Traverse "Competes_with"
competitors = []
for co in companies:
    if co in graph:
        comps = [target for rel, target in graph[co] if rel == "Competes_with"]
        competitors.extend(comps)

print(f"Elon -> Companies: {companies}")
print(f"Companies -> Competitors: {competitors}")
print("Vector search would struggle to hop from Elon to BYD without explicit text linking them.")
```

## Conclusion
Graphs excel at 'Multi-hop Reasoning'.
