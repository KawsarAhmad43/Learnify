# Solution: Average Pooling

**Objective**: Implement Average Pooling.

```python
import numpy as np

def avg_pool(image, pool_size=2, stride=2):
    h, w = image.shape
    out_h = (h - pool_size) // stride + 1
    out_w = (w - pool_size) // stride + 1
    
    output = np.zeros((out_h, out_w))
    
    for i in range(out_h):
        for j in range(out_w):
            curr_h = i * stride
            curr_w = j * stride
            window = image[curr_h : curr_h+pool_size, curr_w : curr_w+pool_size]
            # Capture Mean
            output[i, j] = np.mean(window)
            
    return output

img = np.array([
    [1, 3, 2, 4],
    [5, 6, 1, 2],
    [8, 2, 1, 0],
    [1, 0, 7, 5]
])

pooled = avg_pool(img)
print("Ave Pooled:\n", pooled)
```
