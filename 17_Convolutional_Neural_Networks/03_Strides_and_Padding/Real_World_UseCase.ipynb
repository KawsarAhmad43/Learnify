# Real World Use Case: Object Detection (YOLO)

**Scenario**: You are detecting cars in a video. 
*   Input: 416x416 image.
*   Output: Grid of 13x13 cells (where each cell predicts if there is a car center).

**Goal**: You must design the Stride sequence carefully to ensure 416 divides down perfectly to 13.

```python
def simulate_network(input_size):
    current = input_size
    print(f"Input: {current}")
    
    # 5 Layers of Stride 2 (Factor of 32 reduction)
    for i in range(1, 6):
        # (W - 3 + 2*1) / 2 + 1 = W/2
        current = int((current - 3 + 2*1) / 2 + 1)
        print(f"Layer {i} (Stride 2): {current}")
    
    return current

# Check for standard sizes
final_grid = simulate_network(416)
print(f"Final Grid Size for 416: {final_grid} (Target 13? {'Yes' if final_grid==13 else 'No'})")

# Check for bad size
final_bad = simulate_network(400)
print(f"Final Grid Size for 400: {final_bad} (Target 13? {'Yes' if final_bad==13 else 'No'})")
```

## Conclusion
If you pick an input size like 400, you end up with 12.5 (12) grid cells, which misaligns your bounding boxes. 
This is why YOLO requires inputs to be multiples of 32 (320, 416, 608).
