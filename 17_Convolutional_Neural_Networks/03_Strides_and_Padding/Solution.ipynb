# Solution: Output Shape Calculation

**Objective**: Confirm math.

```python
W = 28
K = 5
P = 2
S = 1

output = (W - K + 2 * P) / S + 1
print(f"Output Size: {output}")
# (28 - 5 + 4) / 1 + 1
# 27 + 1 = 28

print(f"Is Same? {output == W}")
```

## Conclusion
Yes, Padding=2 is correct for a 5x5 kernel to maintain dimensions. 
Rule of thumb for Same padding: $P = (K - 1) / 2$.
