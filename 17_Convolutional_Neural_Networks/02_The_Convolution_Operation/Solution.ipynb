# Solution: Diagonal Detection

**Objective**: Detect diagonal lines.

```python
import numpy as np
from scipy.signal import convolve2d
from sklearn.datasets import load_sample_image
import matplotlib.pyplot as plt

# 1. Load
china = load_sample_image("china.jpg")
china_gray = np.mean(china, axis=2)

# 2. Define Diagonal Kernel
# This checks for transition from Top-Left to Bottom-Right
diag_kernel = np.array([
    [2, 0, 0],
    [0, -1, 0],
    [0, 0, -1]
])

# 3. Convolve
out = convolve2d(china_gray, diag_kernel)

# 4. Show
plt.imshow(np.abs(out), cmap='gray')
plt.title("Diagonal Edges")
plt.axis('off')
plt.show()
```
