# Solution: Validation Split

**Objective**: Data Management.

```python
from sklearn.model_selection import train_test_split
import numpy as np

# Mock Data
all_files = [f"img_{i}.jpg" for i in range(100)]
labels = [i % 2 for i in range(100)]

class ImageDataset:
    def __init__(self, files): # Simplified for demo
        self.files = files
    def __len__(self):
        return len(self.files)

# 1. Split Paths
train_files, val_files = train_test_split(all_files, test_size=0.2, random_state=42)

# 2. Create Objects
train_ds = ImageDataset(train_files)
val_ds = ImageDataset(val_files)

# 3. Verify
print(f"Train Size: {len(train_ds)}") # Should be 80
print(f"Val Size: {len(val_ds)}")     # Should be 20
```
