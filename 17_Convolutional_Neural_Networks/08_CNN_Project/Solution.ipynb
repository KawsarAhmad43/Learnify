# Solution: Adding Dropout

**Objective**: Regularize the network.

```python
class RobustCNN:
    def __init__(self):
        self.layers = [
            "Conv2D(32 Filters, 3x3)",
            "ReLU",
            "MaxPooling(2x2)",
            "Flatten",
            "Dropout(0.5)",  # Added
            "Dense(128 Neurons)",
            "ReLU",
            "Dense(10 Classes)",
            "Softmax"
        ]
        
    def summary(self):
        print("--- Robust Model ---")
        for layer in self.layers:
            print(layer)

model = RobustCNN()
model.summary()
```
