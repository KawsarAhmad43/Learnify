# Implementation: Building a CNN Class

**Goal**: Assemble a full network.

```python
import numpy as np

# Mocking a Framework (Keras-style)
class SimpleCNN:
    def __init__(self):
        self.layers = [
            "Conv2D(32 Filters, 3x3)",
            "ReLU",
            "MaxPooling(2x2)",
            "Flatten",
            "Dense(128 Neurons)",
            "ReLU",
            "Dense(10 Classes)",
            "Softmax"
        ]
        
    def summary(self):
        print("--- Model Summary ---")
        for layer in self.layers:
            print(layer)

model = SimpleCNN()
model.summary()
```

## Conclusion
This architecture (Conv-Pool-Dense) is the bread and butter of image classification.
