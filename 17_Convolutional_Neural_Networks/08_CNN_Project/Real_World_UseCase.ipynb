# Real World Use Case: Satellite Imagery

**Scenario**: Classifying land usage (Forest vs Urban) from satellite tiles.
**Constraint**: Satellite images are huge (10,000 x 10,000). You must chop them into tiles (e.g., 256x256) and classify each tile.
**Goal**: Design the logic for tiling.

```python
import numpy as np

def grid_predict(huge_image_size, tile_size):
    h, w = huge_image_size
    rows = h // tile_size
    cols = w // tile_size
    
    print(f"Image Size: {h}x{w}")
    print(f"Tile Size: {tile_size}x{tile_size}")
    print(f"Total Tiles to Classify: {rows * cols}")
    return rows * cols

# 10k x 10k image
total = grid_predict((10000, 10000), 256)

# If inference takes 0.1s per tile...
time = total * 0.1 / 60
print(f"Estimated processing time: {time:.1f} minutes per huge image.")
```
