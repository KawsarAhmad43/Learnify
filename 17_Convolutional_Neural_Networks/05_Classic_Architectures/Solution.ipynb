# Solution: Parameter Calculation

**Objective**: Count Parameters.

```python
def count_conv_params(kernel_size, input_channels, filters):
    # Weights: K * K * C_in * Filters
    weights = kernel_size * kernel_size * input_channels * filters
    # Biases: Filters
    biases = filters
    total = weights + biases
    return total

k = 3
c_in = 3
f = 10

params = count_conv_params(k, c_in, f)
print(f"Total Parameters: {params}")
```
