# Implementation: Augmentation Pipeline

**Goal**: Use libraries to transform images on the fly.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_sample_image
from scipy.ndimage import rotate

# 1. Load Image
img = load_sample_image("flower.jpg")
img = img / 255.0

# 2. Augment Function (Manual)
def augment(image):
    # Random Flip Left-Right
    if np.random.rand() > 0.5:
        image = np.fliplr(image)
        
    # Random Rotate (-20 to 20)
    angle = np.random.uniform(-20, 20)
    # reshape=False keeps output size same (crops corners)
    image = rotate(image, angle, reshape=False, mode='nearest')
    
    return image

# 3. Visualize a Batch
fig, ax = plt.subplots(1, 5, figsize=(15, 3))
for i in range(5):
    aug_img = augment(img)
    ax[i].imshow(aug_img)
    ax[i].axis('off')
plt.show()
```

## Conclusion
From 1 original content, we created 5 "new" training examples. This is standard preparation for training CNNs.
