# Solution: Gradient Tracing

**Objective**: Demonstrate Vanishing vs Healthy Gradients.

```python
grad_vanish = 1.0
grad_resnet = 1.0

# Simulate 10 layers
for i in range(10):
    grad_vanish *= 0.1 # Weight is small
    grad_resnet *= 1.1 # Weight (0.1) + Skip (1.0)

print(f"Vanishing after 10 layers: {grad_vanish:.10f}")
print(f"ResNet after 10 layers: {grad_resnet:.2f}")
```
