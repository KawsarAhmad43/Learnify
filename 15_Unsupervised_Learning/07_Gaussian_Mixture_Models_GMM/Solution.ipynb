# Solution: GMM Iris Generation

```python
from sklearn.datasets import load_iris
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# 1. Fit
X = load_iris().data
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)

# 2. Generate New Data
X_new, y_new = gmm.sample(100)
print(f"Generated {len(X_new)} synthetic flowers.")

# 3. Plot Original vs Generated (in 2D)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(X[:, 0], X[:, 1], alpha=0.6)
axes[0].set_title("Original Data")
axes[1].scatter(X_new[:, 0], X_new[:, 1], c='red', alpha=0.6)
axes[1].set_title("Generative Data")
plt.show()
```
