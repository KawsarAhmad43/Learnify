# Implementation: Isolation Forest

Detecting outliers in synthetic data.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# 1. Generate Data with Outliers
rng = np.random.RandomState(42)
X = 0.3 * rng.randn(100, 2)
X_train = np.r_[X + 2, X - 2]
# Generate some outliers
X_outliers = rng.uniform(low=-4, high=4, size=(20, 2))

X_all = np.r_[X_train, X_outliers]

# 2. Train Isolation Forest
clf = IsolationForest(contamination=0.1) # We guess ~10% is noise
clf.fit(X_all)
y_pred = clf.predict(X_all)

# 3. Plot
plt.figure(figsize=(8, 6))
plt.scatter(X_all[:, 0], X_all[:, 1], c=y_pred, cmap='Paired')
plt.title("Isolation Forest Outcomes (Blue=Normal, Brown=Outlier)")
plt.show()
```
