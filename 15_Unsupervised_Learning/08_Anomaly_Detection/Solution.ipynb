# Solution: Novelty Detection with Digits

```python
from sklearn.datasets import load_digits
from sklearn.ensemble import IsolationForest
import numpy as np

# 1. Data
digits = load_digits()
X = digits.data
y = digits.target

# Train on 0s only
X_train = X[y == 0]
# Test on 0s and 1s
X_test_normal = X[y == 0][:20] # Some withheld 0s
X_test_anomaly = X[y == 1][:20] # Observations of 1s
X_test = np.concatenate([X_test_normal, X_test_anomaly])

# 2. Train One-Class Model (Isolation Forest)
clf = IsolationForest(contamination=0.01) # Approx 0% is anomaly in training
clf.fit(X_train)

# 3. Predict
preds = clf.predict(X_test)

# 4. Evaluate
print("Predictions (1=Normal, -1=Anomaly):")
print(preds)
print(f"\nCorrectly Flagged Anomalies (should be 20): {np.sum(preds[20:] == -1)}")
```
