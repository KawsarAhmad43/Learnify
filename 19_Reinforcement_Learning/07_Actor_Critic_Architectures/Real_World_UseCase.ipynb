# Real World Use Case: OpenAI Five (Dota 2)

**Scenario**: 5 AI bots controlling heroes.
**Architecture**: Massive Actor-Critic (PPO based).
**Observation**: Screen, Map, Health, Cooldowns (20,000 numbers).
**Critic's Job**: Predict probability of winning the game from current state.
**Actor's Job**: Pick next spell to cast.

```python
# Mock Win Probability

win_prob_now = 0.55
action = "Cast Fireball"
win_prob_next = 0.60

delta = win_prob_next - win_prob_now

print(f"Move: {action}")
print(f"Win Probability Change: {delta:+.2f}")

if delta > 0:
    print("Critic says: Good job! Do that again.")
else:
    print("Critic says: You messed up.")
```

## Conclusion
By having a dense signal ("Win probability went up 1%"), the agent learns much faster than waiting 45 minutes for the "Game Over" screen.
