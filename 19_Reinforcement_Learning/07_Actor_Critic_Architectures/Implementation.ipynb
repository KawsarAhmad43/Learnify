# Implementation: Calculating Advantage

**Goal**: Reduce variance by introducing a baseline.

```python
# Scenario:
# State S. 
# Critic says V(S) = 10.0.
# Agent takes Action A.
# Reward R = 2.0.
# Next State S'. Critic says V(S') = 9.0.
# Gamma = 0.99.

v_s = 10.0
reward = 2.0
v_next = 9.0
gamma = 0.99

# 1. TD Target (What actually happened)
td_target = reward + gamma * v_next

# 2. Advantage (Difference from expectation)
advantage = td_target - v_s

print(f"Target: {td_target:.2f}")
print(f"Advantage: {advantage:.2f}")

# Interpretation:
# If Advantage > 0: Action A was better than average. Encourage it.
# If Advantage < 0: Action A was worse than average. Discourage it.
```

## Conclusion
The Critic (Value Function) acts as a baseline. Instead of asking "Did I win?", we ask "Did I do better than usual?".
