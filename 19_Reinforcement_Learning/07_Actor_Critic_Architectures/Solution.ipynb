# Solution: Shared Architecture

**Objective**: Defining the model structure.

```python
# Pseudo-PyTorch Code
print("class ActorCritic(nn.Module):")
print("    def __init__(self):")
print("        super().__init__()")
print("        self.shared = nn.Linear(4, 64)")
print("        self.actor = nn.Linear(64, 2)")
print("        self.critic = nn.Linear(64, 1)")
print("")
print("    def forward(self, x):")
print("        x = F.relu(self.shared(x))")
print("        probs = F.softmax(self.actor(x), dim=-1)")
print("        value = self.critic(x)")
print("        return probs, value")
```
