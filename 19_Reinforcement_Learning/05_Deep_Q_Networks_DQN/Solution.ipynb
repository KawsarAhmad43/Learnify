# Solution: Epsilon Decay

**Objective**: Visualize schedule.

```python
epsilon = 1.0
decay = 0.995
min_epsilon = 0.01

for episode in range(1000):
    epsilon = max(min_epsilon, epsilon * decay)
    if epsilon < 0.1:
        print(f"Crossed 0.1 at Episode {episode}")
        break

print(f"Final Epsilon: {epsilon:.4f}")
```
