# Solution: Reward Shaping

**Objective**: Engineering the signal.

```python
def get_reward(dist, collision):
    if collision:
        return -100
    
    # Shaping: Higher reward for lower distance
    # Max reward (at dist 0) = 10
    # At dist 10 = 0
    proximity_reward = max(0, 10 - dist)
    
    time_penalty = -1
    
    return proximity_reward + time_penalty

# Test Cases
print(f"Collision: {get_reward(5, True)}")   # Expect -100
print(f"Perfect Park: {get_reward(0, False)}") # Expect 9 (10 - 1)
print(f"Far Away: {get_reward(10, False)}")    # Expect -1 (0 - 1)"""
```
