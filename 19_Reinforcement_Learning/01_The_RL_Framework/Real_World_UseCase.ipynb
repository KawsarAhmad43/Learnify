# Real World Use Case: Smart Thermostat

**Scenario**: Saving energy while keeping a room comfortable.
**State**: Current Temperature.
**Action**: Heater On (1) or Off (0).
**Reward**: 
*   -1 for every minute Heater is On (Energy Cost).
*   -10 if Temperature is < 68F or > 75F (Discomfort).
**Goal**: Find the balance.

```python
class ThermostatEnv:
    def __init__(self):
        self.temp = 70
        
    def step(self, action):
        # Physics
        if action == 1: 
            self.temp += 1
            cost = -1
        else: 
            self.temp -= 1
            cost = 0
            
        # Comfort Penalty
        discomfort = 0
        if self.temp < 68 or self.temp > 75:
            discomfort = -10
            
        return self.temp, cost + discomfort

env = ThermostatEnv()
print(f"Start Temp: {env.temp}")

# Scenario: Leaving heater on too long
total_reward = 0
for _ in range(10):
    t, r = env.step(1) # Always On
    total_reward += r
    print(f"Temp {t}, Reward {r}")

print(f"Total Return: {total_reward}")
```

## Conclusion
The agent got penalized heavily once temperature crossed 75. It needs to learn to switch off.
