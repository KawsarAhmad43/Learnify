# Solution: Backward Return Calculation

**Objective**: Compute Returns.

```python
rewards = [0, 0, 10]
gamma = 0.9

# Accumulator
G = 0
returns = []

# Go backwards
for r in reversed(rewards):
    G = r + gamma * G
    returns.insert(0, G)

for i, g in enumerate(returns):
    print(f"Time {i}: Return G = {g:.2f}")
```
