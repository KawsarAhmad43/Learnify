# Implementation: Monte Carlo Estimator

**Goal**: Estimate the value of a State by averaging returns.

```python
import numpy as np

# 1. Mock History of Episodes from State S
# Episode 1: S -> ... -> Return 10
# Episode 2: S -> ... -> Return -5
# Episode 3: S -> ... -> Return 12
returns_log = [10, -5, 12, 8, 9, -2, 10]

def monte_carlo_value(returns, alpha=0.1):
    # Method 1: True Average
    true_avg = np.mean(returns)
    print(f"True Average Value: {true_avg:.2f}")
    
    # Method 2: Incremental (Moving Average)
    # Good for non-stationary environments or online learning
    value = 0
    for r in returns:
        value = value + alpha * (r - value)
        # New Value = Old Value + Learning Rate * Error
    print(f"Incremental Value (alpha={alpha}): {value:.2f}")

monte_carlo_value(returns_log)
```

## Conclusion
Monte Carlo is simply averaging experience. If you play Blackjack 1 million times and hit on 16, and lose 60% of the time, you learn that 16 is a "Bad State".
