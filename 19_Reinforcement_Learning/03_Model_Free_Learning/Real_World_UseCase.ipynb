# Real World Use Case: Blackjack Strategy

**Scenario**: Learning Basic Strategy in Casinos.
**State**: Your sum (e.g., 14) + Dealer's showing card (e.g., Ace).
**Action**: Hit or Stick.
**Model**: We don't know the deck order (It's shuffled). We can't calculate exact probabilities easily.
**Method**: Simulate 100 million hands. Count wins/losses for every (State, Action) pair.

```python
import numpy as np

# Mock Simulations for State: (Player 16, Dealer 10)
# Action: HIT
results_hit = np.random.choice([-1, 1], size=1000, p=[0.7, 0.3]) # Lose 70%

# Action: STICK
results_stick = np.random.choice([-1, 1], size=1000, p=[0.75, 0.25]) # Lose 75%

value_hit = np.mean(results_hit)
value_stick = np.mean(results_stick)

print(f"Value of HIT: {value_hit:.3f}")
print(f"Value of STICK: {value_stick:.3f}")
print("Both are bad, but HIT is slightly less bad. So the policy says HIT.")
```

## Conclusion
This is how Casino charts are generated. Pure Monte Carlo simulation.
