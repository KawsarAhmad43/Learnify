# Solution: Bellman Update

**Objective**: Calculate V(S).

```python
v_n1 = 50
v_n2 = 20
gamma = 0.9
reward = -1

option1 = reward + gamma * v_n1
option2 = reward + gamma * v_n2

v_s = max(option1, option2)

print(f"Option 1 Value: {option1}")
print(f"Option 2 Value: {option2}")
print(f"New State Value: {v_s}")
print(f"Choice: {'Neighbor 1' if option1 > option2 else 'Neighbor 2'}")
```
