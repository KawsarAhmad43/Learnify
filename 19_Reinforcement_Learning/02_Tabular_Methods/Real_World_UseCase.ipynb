# Real World Use Case: Warehouse Robots

**Scenario**: Kiva robots moving shelves in an Amazon center.
**Setup**: The floor is a Grid.
**Constraint**: We know exactly where the shelves are and where the packing stations are.
**Solution**: Since the map is known and static, we don't need Deep Learning. We use **Path Planning** (A* or Value Iteration) to find the shortest path.

```python
import numpy as np

# 5x5 Grid.
# 0 = Empty, 1 = Value (High means close to goal)
grid = np.zeros((5, 5))

# Goal is at (4, 4)
goal = (4, 4)
grid[goal] = 100

# Visualize Heatmap logic (Manhattan distance manually for demo)
for r in range(5):
    for c in range(5):
        dist = abs(goal[0] - r) + abs(goal[1] - c)
        if dist > 0:
            grid[r, c] = 100 - (dist * 10)

print("Value Map (Robot uses this to climb the hill):")
print(grid)
```

## Conclusion
The robot just looks at its neighbors and moves to the square with the highest number. No "learning" required at runtime.
