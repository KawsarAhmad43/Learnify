# Real World Use Case: Robotics Control

**Scenario**: Teaching a robot dog to walk.
**Action Space**: Torque for 12 motors (Continuous). Actions are not just "On/Off", but any float value.
**Why PG?**: DQN can't handle infinite action choices. Policy Gradient outputs the *Mean* and *Variance* of a Gaussian distribution for every motor.
*   Motor 1: $N(0.5, 0.1)$ 
*   Agent samples from this distribution.

```python
import numpy as np

# Neural Net outputs Mean and StdDev for a motor torque
mu = 2.5 # Torque amount
sigma = 0.5 # Exploration / Uncertainty

# Sample an action
action = np.random.normal(mu, sigma)
print(f"Robot Motor Torque Applied: {action:.4f} Nm")

# If this falls over (Return = -100), we shift 'mu' away from this action.
```
