# Implementation: REINFORCE Logic

**Goal**: Conceptualize the gradient update.

```python
import numpy as np

# 1. Mock Policy Output (Softmax)
# Actions: [Left, Right]
probs = np.array([0.2, 0.8]) 
# Model currently prefers Right heavily.

# 2. Action Taken
action_taken = 0 # Agent explored 'Left' (despite low prob)

# 3. Result
return_G = 100 # It turned out 'Left' was AMAZING.

# 4. Gradient update logic
# We want to INCREASE the probability of index 0.
# Loss = -log(prob[action]) * Return
loss = -np.log(probs[action_taken]) * return_G

print(f"Prob of Action Left: {probs[0]}")
print(f"Log Prob: {np.log(probs[0]):.4f}")
print(f"Loss (to minimize): {loss:.4f}")

# In PyTorch, calling loss.backward() would increase probs[0] for next time.
```

## Conclusion
REINFORCE is elegant but high variance. If one lucky random move wins the game, we reinforce ALL previous moves, even the bad ones (Noise).
