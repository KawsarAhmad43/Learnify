# Solution: Log Identity

**Objective**: Math verification.

```python
import numpy as np

p1 = 0.5
p2 = 0.2

product = p1 * p2
log_prod = np.log(product)

sum_logs = np.log(p1) + np.log(p2)

print(f"Log(p1 * p2): {log_prod}")
print(f"Log(p1) + Log(p2): {sum_logs}")
print(f"Difference: {abs(log_prod - sum_logs):.10f}")
```
