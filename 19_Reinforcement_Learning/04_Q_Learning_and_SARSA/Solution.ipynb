# Solution: Q vs SARSA Logic

**Objective**: Demonstrate update difference.

```python
q_val = 10
reward = -1
alpha = 0.5
gamma = 1.0

# Scenario: Next state is the EDGE of a cliff.
# Optimal Action (Right) -> Value 100 (Goal)
# Actual Random Action (Down) -> Value -100 (Death)

max_q_next = 100 # Q-Learning sees the potential 100
actual_q_next = -100 # SARSA sees the impending doom

new_q_learning = q_val + alpha * (reward + gamma * max_q_next - q_val)
new_sarsa = q_val + alpha * (reward + gamma * actual_q_next - q_val)

print(f"Q-Learning Estimate: {new_q_learning} (Optimistic)")
print(f"SARSA Estimate: {new_sarsa} (Realistic/Safe)")
```
