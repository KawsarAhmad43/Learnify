# Real World Use Case: RLHF (ChatGPT)

**Scenario**: Aligning GPT-4.
**Step 1**: SFT (Supervised Fine Tuning). Humans write answers, model copies them.
**Step 2**: Reward Model. Humans rank 4 answers (A > B > C > D). Train a model to predict scores.
**Step 3**: PPO. Train the LLM to maximize the Reward Model's score.
**Why PPO?**: We don't want the model to hack the reward function and start outputting gibberish that statistically looks good but means nothing. The PPO clip keeps it close to the SFT model (fluent English).

```python
# Conceptual PPO update in Language Models

token_prob_old = 0.1 # Probability of saying "Useful"
token_prob_new = 0.15 # We want to verify update size

kl_divergence = np.log(token_prob_new / token_prob_old)
print(f"KL Penalty: {kl_divergence:.4f}")

# If KL Divergence gets too high, we penalize the reward.
```

## Conclusion
LLM training is just RL with a "Don't change too much" penalty.
