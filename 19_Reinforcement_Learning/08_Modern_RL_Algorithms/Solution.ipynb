# Solution: Clipping

**Objective**: Manual Calculation.

```python
import numpy as np

epsilon = 0.2
lower = 1 - epsilon
upper = 1 + epsilon

case1_ratio = 0.55 / 0.5
case2_ratio = 0.8 / 0.5

clipped1 = np.clip(case1_ratio, lower, upper)
clipped2 = np.clip(case2_ratio, lower, upper)

print(f"Case 1 Ratio: {case1_ratio} -> Clipped: {clipped1}")
print(f"Case 2 Ratio: {case2_ratio} -> Clipped: {clipped2}")
```
