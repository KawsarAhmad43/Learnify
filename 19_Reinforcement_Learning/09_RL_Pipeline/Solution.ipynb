# Solution: Reward Clipping

**Objective**: Data Safety.

```python
class RewardClipWrapper:
    def __init__(self, env=None):
        self.env = env # Mocking composition
        
    def step(self, reward): # Simplified signature for demo
        return max(-1, min(1, reward))

wrapper = RewardClipWrapper()

print(f"Region 100 -> {wrapper.step(100)}")
print(f"Region -50 -> {wrapper.step(-50)}")
print(f"Region 0.5 -> {wrapper.step(0.5)}")
```
