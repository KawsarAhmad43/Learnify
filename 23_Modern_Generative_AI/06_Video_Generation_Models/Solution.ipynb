# Solution: Video Memory

**Objective**: Feasibility.

```python
frames = 96
base_vram = 8 # GB

naive_vram = frames * base_vram

print(f"Naive: {naive_vram} GB")
print("Optimized (Factor 4 compression):", naive_vram / 4, "GB")
print("Still huge. That's why consumer video generators are limited to 2-4 seconds.")
```
