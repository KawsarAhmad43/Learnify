# Real World Use Case: Amazon

**Scenario**: 35% of Amazon's revenue comes from recommendations.
**Algorithm**: Item-to-Item Collaborative Filtering.
**Logic**: 
1.  They don't measure User Similarity (User history changes too fast).
2.  They measure Item Similarity. "Users who bought Shoes also bought Socks".
3.  This relationship is stable over time.
**Result**: Extremely fast lookup table for "Related Products".
