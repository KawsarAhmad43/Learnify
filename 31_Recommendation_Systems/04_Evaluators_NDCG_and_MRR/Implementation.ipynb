# Implementation: Calculating NDCG@5

**Goal**: Measure Ranking Quality.

```python
import numpy as np
from sklearn.metrics import ndcg_score

# 1. True Relevancy (Ground Truth)
# Movie A: 5 stars (Most relevant)
# Movie B: 1 star (Least relevant)
# Movie C: 0 stars (Irrelevant)
# Movie D: 0 stars (Irrelevant)
# Movie E: 0 stars (Irrelevant)
true_relevance = np.asarray([[5, 1, 0, 0, 0]])

# 2. Model Predictions (Scores)
# Scenario 1: Perfect Order
good_scores = np.asarray([[0.9, 0.8, 0.5, 0.3, 0.1]])
# Scenario 2: Terrible Order (Best movie is last)
bad_scores = np.asarray([[0.1, 0.2, 0.3, 0.4, 0.9]])

# 3. Calculate NDCG
print(f"Good NDCG: {ndcg_score(true_relevance, good_scores):.4f}")
print(f"Bad NDCG: {ndcg_score(true_relevance, bad_scores):.4f}")
```

## Conclusion
If you optimize for MSE (Mean Squared Error), your NDCG might still be bad. Optimize for Ranking Loss (Pairwise Ranking).
