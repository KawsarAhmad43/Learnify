# Real World Use Case: Netflix A/B Testing

**Scenario**: Assessing a new model.
**Offline Metric**: The Data Science team sees that NDCG increased by +0.5%.
**Online Metric**: They deploy it to 5% of users. They discover that Watch Time decreased by -2%.
**Why?**: The model was ranking "Clickbait" thumbnails very high (users clicked, so NDCG was high), but the content was bad (users quit instantly).
**Lesson**: NDCG measures *Relevance Prediction*, not *User Satisfaction*. You need both.
