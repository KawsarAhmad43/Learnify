# Solution: Feature Engineering

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Length': [10, 20, 30],
    'Width': [5, 10, 15],
    'Date': ['2023-01-01', '2023-07-04', '2023-12-25']
})

# 1. Interaction
df['Area'] = df['Length'] * df['Width']

# 2. Date Features
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['DayOfWeek'] = df['Date'].dt.day_name()

# 3. Log Transform
df['Log_Area'] = np.log1p(df['Area'])

print("Engineered DataFrame:")
display(df)
```
