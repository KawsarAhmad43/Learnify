# Solution: Handling Missing Data

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Category': ['A', 'B', np.nan, 'A', 'A', 'B'],
    'Score': [10, 20, 10, np.nan, 50, 60],
    'Critical_Feature': [1, 0, 1, 0, np.nan, 1]
})

# 1. Count NaNs
print("Missing Count:")
print(df.isnull().sum())

# 2. Fill Category with Mode
mode_val = df['Category'].mode()[0]
df['Category'] = df['Category'].fillna(mode_val)

# 3. Fill Score with Median
median_val = df['Score'].median()
df['Score'] = df['Score'].fillna(median_val)

# 4. Drop rows where Critical_Feature is NaN
df = df.dropna(subset=['Critical_Feature'])

print("\nCleaned DataFrame:")
display(df)
```
