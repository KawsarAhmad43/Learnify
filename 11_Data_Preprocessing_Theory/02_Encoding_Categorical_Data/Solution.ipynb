# Solution: Encoding Strategies

```python
import pandas as pd

df = pd.DataFrame({
    'City': ['New York', 'Paris', 'New York', 'Tokyo'],
    'Quality': ['Good', 'Excellent', 'Poor', 'Good'],
    'Temperature': [20, 25, 15, 22]
})

# 1. Ordinal Encoding for Quality
quality_map = {'Poor': 0, 'Good': 1, 'Excellent': 2}
df['Quality_Encoded'] = df['Quality'].map(quality_map)

# 2. One-Hot Encoding for City (pd.get_dummies is easier than sklearn for quick analysis)
df_encoded = pd.get_dummies(df, columns=['City'], drop_first=True)

print("Final Encoded DataFrame:")
display(df_encoded)
```
