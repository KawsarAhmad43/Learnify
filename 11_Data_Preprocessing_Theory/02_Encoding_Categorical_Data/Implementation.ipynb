# Implementation: Encoding Categorical Data

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

df = pd.DataFrame({
    'Size': ['Small', 'Medium', 'Large', 'Medium'],
    'Color': ['Red', 'Blue', 'Green', 'Red']
})
print("Original Data:")
display(df)
```

## 1. Ordinal Encoding
For the 'Size' column.

```python
order = [['Small', 'Medium', 'Large']]
ord_enc = OrdinalEncoder(categories=order)

df['Size_Encoded'] = ord_enc.fit_transform(df[['Size']])
print("Ordinal Encoded Size:")
display(df)
```

## 2. One-Hot Encoding
For the 'Color' column.

```python
ohe = OneHotEncoder(sparse=False)
color_encoded = ohe.fit_transform(df[['Color']])

# Create DataFrame from the numpy array
color_df = pd.DataFrame(color_encoded, columns=ohe.get_feature_names_out(['Color']))

print("One-Hot Encoded Colors:")
display(color_df)

# Join back
final_df = pd.concat([df, color_df], axis=1)
display(final_df)
```
