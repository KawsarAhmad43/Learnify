# Solution: Advanced Data Tricks

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Review': ["The product is great but shipping was slow."],
    'Timestamp': ['2023-10-31 23:59:59'],
    'Pixel_Val': [128]
})

# 1. Text
df['Review_Clean'] = df['Review'].str.lower()
df['Word_Count'] = df['Review_Clean'].apply(lambda x: len(x.split()))

# 2. Cyclical Time
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
hour = df['Timestamp'].dt.hour
df['Hour_Sin'] = np.sin(2 * np.pi * hour / 24)
df['Hour_Cos'] = np.cos(2 * np.pi * hour / 24)

# 3. Image Normalization
df['Pixel_Norm'] = df['Pixel_Val'] / 255.0

print("Processed Data:")
display(df)
```
