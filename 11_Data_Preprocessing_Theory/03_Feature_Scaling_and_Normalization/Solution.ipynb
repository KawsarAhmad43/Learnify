# Solution: Feature Scaling

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

df = pd.DataFrame({
    'Age': [20, 30, 40, 50, 60],
    'Salary': [20000, 30000, 40000, 50000, 1000000]
})

# 1. StandardScaler
std = StandardScaler()
df_std = pd.DataFrame(std.fit_transform(df), columns=['Age', 'Salary'])
print("Standard Scaled Stats:")
display(df_std.describe().round(2))

# 2. MinMax
minmax = MinMaxScaler()
df_minmax = pd.DataFrame(minmax.fit_transform(df), columns=['Age', 'Salary'])
print("\nMinMax Range:")
print(f"Max: {df_minmax.max().max()}, Min: {df_minmax.min().min()}")

# 3. Discussion
print("\nDiscussion: The Millionaire (Salary=1M) is a huge outlier.")
print("StandardScaler is skewed by the mean. MinMaxScaler squashes everyone else into a tiny range [0, 0.03].")
print("RobustScaler would be the best choice here!")
```
