# Real World Use Case: Predictive Keyboard

**Task A**: Standard English (trained on Books).
**Task B**: User Specific Slang (trained on your texts).
**Task C**: New Pop Culture words (trained on Twitter trends).

**Scenario**: You type "lol". 
*   If the model forgot Task A, it might not know what "the" means anymore.
*   We need it to learn "rizz" (Task C) without forgetting "hello" (Task A).
**Method**: On-device Replay. The phone keeps a tiny database of your past 1000 messages and mixes them in whenever it updates the model overnight.

```python
# Replay Buffer Mix
new_data = ["I got that rizz", "no cap based"]
replay_buffer = ["Hello sir", "How are you?"]

training_batch = new_data + replay_buffer
print(f"Training Batch: {training_batch}")
```

## Conclusion
If you don't use replay, after one week of talking to your Gen-Z nephew, your keyboard would stop predicting formal english for your Boss.
