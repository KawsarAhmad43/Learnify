# Solution: Drift Calculation

**Objective**: Measure change.

```python
import numpy as np

w_old = np.array([0.5, -0.5])
w_new = np.array([0.6, -0.1])

drift = np.sqrt(np.sum((w_new - w_old)**2))
print(f"L2 Drift: {drift:.4f}")

diffs = np.abs(w_new - w_old)
print(f"Per parameter drift: {diffs}")
print(f"Max drift index: {np.argmax(diffs)}")
```
