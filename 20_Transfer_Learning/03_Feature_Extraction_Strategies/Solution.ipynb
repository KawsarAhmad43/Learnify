# Solution: Parameter Counting

**Objective**: Count trainable params.

```python
layers = {
    "conv1": 100,
    "conv2": 200,
    "head": 50
}

requires_grad = {
    "conv1": False,
    "conv2": True, # Unfrozen
    "head": True
}

total_trainable = 0
for name, size in layers.items():
    if requires_grad[name]:
        total_trainable += size

print(f"Total Trainable: {total_trainable}")
print(f"Forward Pass is full cost.")
print(f"Backward Pass is saved for conv1.")
```
