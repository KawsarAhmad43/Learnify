# Solution: Gradient Addition

**Objective**: Manual verification.

```python
x = 1.0
w = 0.0

# Target 1 = 1, Target 2 = 2
grad1 = 2 * (w*x - 1) * x
grad2 = 2 * (w*x - 2) * x

total_grad = grad1 + grad2

print(f"Grad 1: {grad1}")
print(f"Grad 2: {grad2}")
print(f"Combined: {total_grad}")
print("The shared weight learns to satisfy the 'average' demand of both tasks.")
```
