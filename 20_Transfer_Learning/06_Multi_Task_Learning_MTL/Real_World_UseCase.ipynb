# Real World Use Case: Tesla Autopilot (HydraNet)

**Input**: 8 Cameras.
**Backbone**: RegNet (Massive feature extractor).
**Heads**:
1.  **Lane Detection**: Where are the lines?
2.  **Object Detection**: Where are the cars?
3.  **Traffic Lights**: Red or Green?
4.  **Depth Estimation**: How far away?

**Why?**: You can't run 4 separate ResNet50s on a car computer (Latency/Heat). You run 1 Backbone and 4 tiny heads.
**Benefit**: Knowing where the "Car" is helps you know where the "Lane" is (Context).

```python
# Conceptual Loss Combination
loss_lane = 0.5
loss_obj = 2.0
loss_light = 0.1

# Naive Sum
total = loss_lane + loss_obj + loss_light
print(f"Total Loss: {total}")
print("Object Detection dominates the gradient (2.0 vs 0.1). the model will ignore Traffic Lights.")

# Weighted Sum
total_weighted = 1.0*loss_lane + 0.1*loss_obj + 10.0*loss_light
print(f"Weighted Loss: {total_weighted}")
print("Now Traffic Lights matter.")
```

## Conclusion
Balancing these weights (HydraNet tuning) is one of the hardest parts of ML Engineering.
