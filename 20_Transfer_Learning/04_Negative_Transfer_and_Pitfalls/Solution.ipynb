# Solution: Domain Ranking

**Objective**: Evaluating fit.

```python
rank = {
    "CIFAR-10": "Low Risk (Similar classes, just smaller pixels)",
    "Satellite": "Medium Risk (Different viewpoint, but edges/roads are familiar)",
    "Microscope": "High Risk (Completely alien physics/lighting)"
}

for domain, risk in rank.items():
    print(f"{domain}: {risk}")
```
