# Real World Use Case: NLP for Code

**Source**: GPT-3 trained on English Text (Wikipedia, Reddit).
**Target**: Python Code generation.
**Issue**: English grammar and Python syntax conflict.
*   English: "The variable x is equal to 5."
*   Python: `x = 5`
**Negative Transfer**: Early models would output comments instead of code because they prioritized natural language structure.
**Solution**: CODEX (OpenAI). They fine-tuned specifically on GitHub code to override the "English bias".

```python
# Conceptual Example

prompt = "def add(a, b):"

model_english_bias = " This function adds two numbers."
model_code_bias = "\n    return a + b"

print(f"Prompt: {prompt}")
print(f"English Model: {model_english_bias} (Useless)")
print(f"Code Model: {model_code_bias} (Useful)")
```

## Conclusion
Specialized pre-training is better than generic pre-training when domains diverge.
