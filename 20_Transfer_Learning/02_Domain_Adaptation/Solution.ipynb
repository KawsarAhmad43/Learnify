# Solution: Manual Alignment

**Objective**: Shift alignment.

```python
import numpy as np

source_mean = 10.0
target_mean = 15.0

target_batch = np.array([14.0, 15.0, 16.0])

# Calculate offset
shift = target_mean - source_mean

# Correct the batch
adapted_batch = target_batch - shift

print(f"Original Target: {target_batch}")
print(f"Adapted Target: {adapted_batch}")
print(f"New Mean: {np.mean(adapted_batch)} (Matches Source Mean 10)")
```
