# Real World Use Case: Self-Driving Cars

**Source Domain**: Grand Theft Auto V (Game). 
*   Why? Infinite labels, free data, safety.
*   Style: Synthetic, sharp edges, perfect colors.

**Target Domain**: Real Streets (San Francisco).
*   Style: Noise, motion blur, complex texture.

**Problem**: A model trained on GTA V fails on real roads because "Road" looks different.
**Solution**: CycleGAN or DANN.
*   CycleGAN: Translates GTA V images to look like Real images.
*   Train the detector on the "Realified" GTA images.
