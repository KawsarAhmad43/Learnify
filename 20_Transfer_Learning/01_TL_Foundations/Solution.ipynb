# Solution: VGG Surgery

**Objective**: Replace the head.

```python
import torch.nn as nn
from torchvision import models

# 1. Load VGG16 (No weights needed for structural check)
vgg = models.vgg16()

# 2. Check original last layer
original_layer = vgg.classifier[6]
print(f"Original: {original_layer}")

# 3. Replace
in_feats = original_layer.in_features
vgg.classifier[6] = nn.Linear(in_feats, 10)

print(f"New: {vgg.classifier[6]}")

# Verification
print("Full Classifier Block:")
print(vgg.classifier)
```
