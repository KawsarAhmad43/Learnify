# Real World Use Case: Dog Breed Classifier

**Scenario**: You want to identify specific breeds (Golden Retriver vs Labrador).
**Data**: You have only 50 photos of each.
**Problem**: Training a CNN from scratch on 100 photos will overfit 100%.
**Solution**: Use ImageNet weights.
*   ImageNet has 1.2M images.
*   It already has classes for "Golden Retriever" and "Labrador".
*   Even if it didn't, it knows what "Fur", "Snout", and "Ears" look like.
**Method**: We assume the Source Domain (ImageNet) is very close to Target Domain (Your Dogs).

```python
# Conceptual Workflow
print("1. Download ResNet50 (Pre-trained).")
print("2. Freeze all layers (Do not destroy the 'knowledge').")
print("3. Remove the last layer (1000 classes).")
print("4. Add a new layer (2 classes).")
print("5. Train ONLY the new layer on your 100 photos.")
print("6. Result: High accuracy with tiny data.")
```
