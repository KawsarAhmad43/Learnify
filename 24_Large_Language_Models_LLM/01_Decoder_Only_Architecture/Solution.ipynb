# Solution: Temperature

**Objective**: Visualization.

```python
import numpy as np

logits = np.array([5.0, 3.0, 0.1])

def sample(logits, T):
    exps = np.exp(logits / T)
    probs = exps / np.sum(exps)
    return probs

print(f"T=0.1: {sample(logits, 0.1)} (Sharp)")
print(f"T=2.0: {sample(logits, 2.0)} (Flat)")
```
