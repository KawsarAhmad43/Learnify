# Real World Use Case: Llama 3

**Scenario**: Assessing Llama 3 8B.
**Fact**: It was trained on 15 Trillion tokens.
**Chinchilla Optimal**: $8 	imes 20 = 160$ Billion.
**Overtraining Ratio**: $15,000 / 160 \approx 93x$.
**Why?**: 
*   Training is a one-time cost ($C_{train}$).
*   Inference is a recurring cost ($C_{infer}$).
*   If millions of people use the model, optimizing for *Inference Efficiency* (smaller model) is worth paying a massive penalty in *Training Inefficiency* (training it way longer than optimal).
**Result**: Llama 3 8B outperforms GPT-3 (175B) despite being 20x smaller.
