# Implementation: Chinchilla Calculator

**Goal**: Plan a training run.

```python
def chinchilla_optimal_tokens(params_billions):
    return params_billions * 20

def training_flops(params_billions, tokens_billions):
    # Approximate FLOPs = 6 * N * D
    # 6 FLOPs per token per parameter (Forward + Backward pass)
    flops = 6 * (params_billions * 1e9) * (tokens_billions * 1e9)
    return flops

# Scenario: I have a 10B param model
N = 10 
optimal_D = chinchilla_optimal_tokens(N)
compute_needed = training_flops(N, optimal_D)

print(f"Model Size: {N}B")
print(f"Optimal Tokens: {optimal_D}B")
print(f"Compute Needed: {compute_needed:.2e} FLOPs")
print("If you use A100 (312 TFLOPS), this takes:", compute_needed / (312 * 1e12) / 3600, "hours on 1 GPU.")
```
