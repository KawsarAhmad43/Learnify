# Implementation: RoPE Logic

**Goal**: Rotate vectors.

```python
import numpy as np

def rotate(x, theta):
    # 2D Rotation matrix
    # [cos -sin]
    # [sin  cos]
    c, s = np.cos(theta), np.sin(theta)
    return np.array([x[0]*c - x[1]*s, x[0]*s + x[1]*c])

# Vector at Pos 0
vec = np.array([1.0, 0.0])

# Pos 1 (Rotated 10 degrees)
vec_pos1 = rotate(vec, np.deg2rad(10))

# Pos 2 (Rotated 20 degrees)
vec_pos2 = rotate(vec, np.deg2rad(20))

# Dot Product (Similarity) depends only on relative angle
print(f"Dot(Pos1, Pos2): {np.dot(vec_pos1, vec_pos2):.4f}")

# Pos 10 and Pos 11 (Rotated 100 and 110)
vec_pos10 = rotate(vec, np.deg2rad(100))
vec_pos11 = rotate(vec, np.deg2rad(110))

print(f"Dot(Pos10, Pos11): {np.dot(vec_pos10, vec_pos11):.4f}")
print("The attention score is identical because the relative distance (1 step) is the same.")
```

## Conclusion
RoPE captures relative distance naturally via rotation.
