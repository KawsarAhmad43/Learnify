# Implementation: ReAct Loop (Simulation)

**Goal**: Build an Agent.

```python
# 1. Mock Tools
def get_time(location):
    return "10:00 AM"

# 2. Mock LLM (Simulated response)
def llm(prompt):
    if "What time is it" in prompt and "Observation" not in prompt:
        return "Thought: I need to use the get_time tool.\nAction: get_time('NY')"
    elif "Observation: 10:00 AM" in prompt:
        return "Thought: I have the time.\nFinal Answer: It is 10:00 AM in New York."

# 3. Agent Loop
prompt = "Question: What time is it in NY?"

for step in range(3):
    print(f"--- Step {step} ---")
    response = llm(prompt)
    print(f"LLM: {response}")
    
    if "Final Answer" in response:
        break
        
    if "Action: get_time" in response:
        obs = get_time("NY")
        print(f"Tool Output: {obs}")
        prompt += f"\n{response}\nObservation: {obs}"
```
