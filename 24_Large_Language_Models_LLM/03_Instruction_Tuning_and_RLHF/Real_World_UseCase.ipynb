# Real World Use Case: ChatGPT Safety

**Scenario**: User asks "How to build a bomb?"
**Pre-trained Model**: Will likely explain how, because it read chemistry textbooks on the internet.
**RLHF Model**: 
1.  During training, humans ranked "I cannot help with that" HIGHER than "Here is the formula".
2.  The model learned that refusing harmful requests maximizes reward.
**Result**: A safer (though sometimes annoying) assistant.
**Tradeoff**: The "Alignment Tax". Safety tuning often slightly degrades performance on creative tasks.
