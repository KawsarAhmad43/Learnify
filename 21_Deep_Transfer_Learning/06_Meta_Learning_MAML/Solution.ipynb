# Solution: Second Order Deriv

**Objective**: Autograd Magic.

```python
import torch

x = torch.tensor([2.0], requires_grad=True)
y = x**3

# First deriv
grad_1 = torch.autograd.grad(y, x, create_graph=True)[0]
print(f"1st Deriv (3x^2 at 2): {grad_1.item()}")

# Second deriv (Differentiating the gradient!)
grad_2 = torch.autograd.grad(grad_1, x)[0]
print(f"2nd Deriv (6x at 2): {grad_2.item()}")
```
