# Solution: Temperature Softmax

**Objective**: Math.

```python
import numpy as np

logits = np.array([2.0, 4.0])

def softmax(x, T=1.0):
    exps = np.exp(x / T)
    return exps / np.sum(exps)

p_t1 = softmax(logits, T=1)
p_t4 = softmax(logits, T=4)

print(f"T=1: {p_t1} (Sharp, almost 90% confidence)")
print(f"T=4: {p_t4} (Soft, only 60% confidence)")
```
