# Real World Use Case: DistilBERT

**Scenario**: You want to run Sentiment Analysis on a Smartphone.
**Problem**: BERT-Base is 440MB and slow (100ms inference). A user typing a review feels the lag.
**Solution**: DistilBERT.
1.  Take BERT-Base (12 Layers).
2.  Create a Student (6 Layers).
3.  Initialize Student with every *other* layer of Teacher.
4.  Train Student to mimic Teacher's output on Wikipedia.
**Result**: 
*   Size: 40% Smaller.
*   Speed: 60% Faster.
*   Accuracy: 97% of Teacher.

```python
print("Latency (BERT): 100ms")
print("Latency (DistilBERT): 40ms")
print("User Experience: Seamless vs Laggy.")
```
