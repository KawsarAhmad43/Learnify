# Real World Use Case: Multi-Tenant Llama 2

**Scenario**: You run a SaaS platform using Llama-2-70B.
**Client A**: Wants a "Pirate Bot".
**Client B**: Wants a "Customer Service Bot".
**Strategy**:
1.  Keep **One** frozen Llama-2 (140GB VRAM) loaded in GPU.
2.  Train a tiny LoRA adapter for Client A (10MB).
3.  Train a tiny LoRA adapter for Client B (10MB).
**Runtime**:
*   Request from A -> `Run(Llama + LoRA_A)`.
*   Request from B -> `Run(Llama + LoRA_B)`.
**Result**: You serve millions of custom personas from a single backbone.

```python
# Pseudo-code for Hot-Swapping

def serve_request(user_id, prompt):
    base_model_output = run_frozen_backbone(prompt)
    
    if user_id == "PirateConfig":
        adapter_output = run_lora_a(prompt)
    elif user_id == "SupportConfig":
        adapter_output = run_lora_b(prompt)
        
    return base_model_output + adapter_output
```
