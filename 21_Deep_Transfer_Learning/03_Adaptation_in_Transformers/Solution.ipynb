# Solution: LoRA Parameter Count

**Objective**: Math.

```python
dim = 1000
rank = 4

full_params = dim * dim
lora_params = (dim * rank) + (rank * dim)

ratio = lora_params / full_params
percentage = ratio * 100

print(f"Full Params: {full_params:,}")
print(f"LoRA Params: {lora_params:,}")
print(f"Size: {percentage:.2f}% of original.")

# If 4MB:
print(f"Memory: {4 * ratio:.4f} MB")
```
