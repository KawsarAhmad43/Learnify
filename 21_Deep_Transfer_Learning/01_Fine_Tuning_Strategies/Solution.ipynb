# Solution: LR Calculation

**Objective**: Compute values.

```python
lr_top = 0.01
decay = 2.6

layer_minus_1 = lr_top
layer_minus_2 = layer_minus_1 / decay
layer_minus_3 = layer_minus_2 / decay

print(f"Top Layer LR: {layer_minus_1:.6f}")
print(f"Mid Layer LR: {layer_minus_2:.6f}")
print(f"Bot Layer LR: {layer_minus_3:.6f}")
```
