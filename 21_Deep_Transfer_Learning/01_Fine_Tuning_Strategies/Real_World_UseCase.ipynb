# Real World Use Case: Legal Text Classifier (ULMFiT)

**Scenario**: Classifying "Commercial Contracts" vs "NDA" vs "Lease".
**Source**: BERT (Trained on Wikipedia).
**Problem**: Wikipedia English is very different from Legalese ("Herein", "Thereafter", "Force Majeure").
**Strategy (ULMFiT)**:
1.  **Domain Adaptation**: Fine-tune *only* the Language Model (Next word prediction) on Legal texts. No classification yet. Just get BERT to learn "Legal English".
2.  **Fine-Tune Classifier**: Add the classifier head. Use Gradual Unfreezing and Triangular Learning Rates.
**Result**: 99% accuracy with only 100 labeled contracts.

```python
# Conceptual Data
wiki_text = "The cat sat on the mat."
legal_text = "The obligations contained herein shall survive termination."

print(f"Source Domain: {wiki_text}")
print(f"Target Domain: {legal_text}")
print("Step 1: Teach model that 'herein' is a common word.")
print("Step 2: Teach model that 'termination' implies end of contract.")
```

## Conclusion
This 3-step process (General Pretraining -> Domain Fine-tuning -> Task Fine-tuning) is the gold standard in NLP.
