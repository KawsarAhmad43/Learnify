# Implementation: Model Card Generator

**Goal**: Automated documentation.

```python
class ModelCard:
    def __init__(self, name, parent_model, fine_tuning_data, architecture_changes):
        self.name = name
        self.parent = parent_model
        self.data = fine_tuning_data
        self.arch = architecture_changes
        
    def generate_markdown(self):
        md = f"# Model Card: {self.name}\n"
        md += f"## Family Tree\n"
        md += f"*   **Base Model**: {self.parent}\n"
        md += f"*   **Architecture**: {self.arch}\n"
        md += f"## Training Details\n"
        md += f"*   **Dataset**: {self.data}\n"
        return md

card = ModelCard(
    name="MedLlama-Tiny",
    parent_model="Llama-3-70B (Instruct)",
    fine_tuning_data="PubMed 100k (QA Pairs)",
    architecture_changes="Distilled to 1B, Quantized to Int8"
)

print(card.generate_markdown())
```

## Conclusion
Every model in your registry should have this file.
