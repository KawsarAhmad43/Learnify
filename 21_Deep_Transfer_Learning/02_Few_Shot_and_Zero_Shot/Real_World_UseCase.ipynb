# Real World Use Case: Face ID (One-Shot Learning)

**Scenario**: Unlocking your phone.
**Problem**: The Neural Network (Siamese Network) was trained on millions of faces at Apple HQ.
**But**: It has NEVER seen your face.
**Setup**: When you buy the phone, you take 5 photos (5-Shot Support Set). The phone calculates your `MyFace_Prototype` vector and saves it (Encrypted).
**Runtime**: When you look at the phone, it generates a query vector. If `dist(Query, MyFace_Prototype) < threshold`, it unlocks.

```python
threshold = 0.6
diff = 0.23 # Calculated distance

if diff < threshold:
    print("Unlock Phone")
else:
    print("Ask for Passcode")
```

## Conclusion
This is why you don't need to "Retrain" your iPhone overnight to recognize you.
