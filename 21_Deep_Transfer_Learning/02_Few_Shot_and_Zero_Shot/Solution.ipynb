# Solution: Distance Calculation

**Objective**: Manual verification.

```python
import numpy as np

A = np.array([0, 0])
B = np.array([10, 0])
C = np.array([0, 10])
Q = np.array([2, 2])

dA = np.linalg.norm(Q - A)
dB = np.linalg.norm(Q - B)
dC = np.linalg.norm(Q - C)

print(f"Dist A: {dA:.2f}")
print(f"Dist B: {dB:.2f}")
print(f"Dist C: {dC:.2f}")
print(f"Winner: Class {'ABC'[np.argmin([dA, dB, dC])]})")
```
